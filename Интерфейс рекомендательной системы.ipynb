```python

nltk.download('stopwords')

# Инициализация морфологического анализатора и стоп-слов
morph = MorphAnalyzer()
russian_stopwords = stopwords.words('russian')
additional_stopwords = ['тд', 'тп', 'тк', 'зарплата', 'вакансия', 'работодатель', 
                       'компания', 'работ', 'работа', 'обязанность', 'требование']
russian_stopwords.extend(additional_stopwords)

def preprocess_text(text):
    """Функция для предобработки текста"""
    if not isinstance(text, str):
        return ""
    
    # Удаление HTML-тегов
    text = re.sub(r'<[^>]+>', ' ', text)
    
    # Удаление спецсимволов и цифр
    text = re.sub(r'[^а-яА-ЯёЁa-zA-Z\-]', ' ', text)
    
    # Приведение к нижнему регистру
    text = text.lower()
    
    # Токенизация
    words = text.split()
    
    # Лемматизация и удаление стоп-слов
    processed_words = []
    for word in words:
        if len(word) > 2 and word not in russian_stopwords:
            parsed_word = morph.parse(word)[0]
            lemma = parsed_word.normal_form
            processed_words.append(lemma)
    
    return ' '.join(processed_words)

def prepare_dataset(data):
    """Подготовка датасета для анализа"""
    # Создаем DataFrame из данных
    df = pd.DataFrame(data)
    
    # Обработка текстовых полей
    text_fields = ['name_desc', 'description_desc', 'branded_description_desc','area.name_desc', 'key_skills','vacancies_url','employer.name','salary.from']
    
    for field in text_fields:
        if field in df.columns:
            if field == 'key_skills':
                # Обработка ключевых навыков (список словарей)
                df[field] = df[field].apply(lambda x: ' '.join([skill['name'] for skill in x]) if isinstance(x, list) else '')
            df[f'processed_{field}'] = df[field].apply(preprocess_text)
    
    # Создаем объединенное текстовое поле для анализа
    df['combined_text'] = ''
    for field in text_fields:
        if f'processed_{field}' in df.columns:
            df['combined_text'] += ' ' + df[f'processed_{field}']
    
    return df

def calculate_similarity(df, target_vacancy_idx):
    """Расчет косинусной схожести между вакансиями"""
    # Инициализация TF-IDF векторайзера
    vectorizer = TfidfVectorizer()
    
    # Преобразование текстов в TF-IDF матрицу
    tfidf_matrix = vectorizer.fit_transform(df['combined_text'])
    
    # Расчет косинусной схожести
    cosine_sim = cosine_similarity(tfidf_matrix)
    
    # Получение схожести для целевой вакансии
    sim_scores = list(enumerate(cosine_sim[target_vacancy_idx]))
    
    # Сортировка вакансий по схожести
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    return sim_scores

def recommend_vacancies(df, sim_scores, top_n=5):
    """Рекомендация вакансий на основе схожести"""
    # Получаем индексы топ-N вакансий
    top_indices = [i[0] for i in sim_scores[1:top_n+1]]
    
    # Возвращаем информацию о рекомендуемых вакансиях
    return df.iloc[top_indices][['id_desc', 'name_desc', 'description_desc']]

def load_and_concatenate_excel_files(directory):
    """Загрузка и объединение всех Excel файлов в указанной директории"""
    all_data = []
    for file_name in os.listdir(directory):
        if file_name.endswith('.xlsx'):
            file_path = os.path.join(directory, file_name)
            df = pd.read_excel(file_path)
            all_data.append(df)
    return pd.concat(all_data, ignore_index=True)

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

import tkinter as tk
from tkinter import ttk, filedialog, messagebox
from tkinter.scrolledtext import ScrolledText
import pandas as pd
import numpy as np
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer

# Function to create and train a simple LSTM model
def create_and_train_lstm_model(vacancies_df):
    # Ensure that the columns contain valid data
    vacancies_df['combined_text'] = vacancies_df['combined_text'].astype(str)
    vacancies_df['name_desc'] = vacancies_df['name_desc'].astype(str)

    # Drop any rows where 'name_desc' is NaN or empty
    vacancies_df = vacancies_df[vacancies_df['name_desc'].notna() & (vacancies_df['name_desc'] != '')]

    texts = vacancies_df['combined_text'].values
    labels = vacancies_df['name_desc'].values

    tokenizer = Tokenizer(num_words=10000)
    tokenizer.fit_on_texts(texts)
    sequences = tokenizer.texts_to_sequences(texts)
    X = pad_sequences(sequences, maxlen=100)
    y = pd.get_dummies(labels).values  # One-hot encoding

    model = Sequential()
    model.add(Embedding(input_dim=10000, output_dim=128))
    model.add(LSTM(64, return_sequences=False))
    model.add(Dense(len(np.unique(labels)), activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    model.fit(X, y, epochs=5, batch_size=32, validation_split=0.2)

    model.save('lstm_model.h5')
    return model, tokenizer

```

```python
import tkinter as tk
from tkinter import ttk, messagebox, filedialog
from tkinter.scrolledtext import ScrolledText
import pandas as pd
import numpy as np
import re
import os
import json
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import tensorflow as tf

class JobSearchSystem:
    def __init__(self, root, vacancies_df):
        self.root = root
        self.root.title("Система поиска вакансий")
        self.root.geometry("900x650")
        
        # Загрузка данных
        self.vacancies_df = vacancies_df
        self.preprocess_data()
        
        # Инициализация моделей
        self.init_models()
        
        # Переменные интерфейса
        self.job_title = tk.StringVar()
        self.resume_text = ""
        self.experience = tk.StringVar(value="Нет опыта")
        self.city = tk.StringVar()
        self.same_field = tk.BooleanVar(value=True)
        self.desired_job = tk.StringVar()
        self.desired_salary = tk.StringVar()
        self.unemployed = tk.BooleanVar(value=False)
        self.employment_type = tk.StringVar(value="Полный день")
        self.work_format = tk.StringVar(value="Не важно")
        
        # Создание интерфейса
        self.create_widgets()
    
    def preprocess_data(self):
        """Предварительная обработка данных"""
        # Заполнение пропущенных значений
        self.vacancies_df['salary.from'] = self.vacancies_df['salary.from'].fillna(0)
        self.vacancies_df['salary.to'] = self.vacancies_df['salary.to'].fillna(0)
        self.vacancies_df['experience.name'] = self.vacancies_df['experience.name'].fillna('Нет опыта')
        
        # Очистка текстовых полей
        text_columns = ['name_desc', 'description', 'combined_text']
        for col in text_columns:
            if col in self.vacancies_df.columns:
                self.vacancies_df[col] = self.vacancies_df[col].fillna('').apply(self.clean_text)
            else:
                # Если колонки нет, создаем ее
                if col == 'combined_text':
                    self.vacancies_df['combined_text'] = (
                        self.vacancies_df.get('name_desc', '') + " " + 
                        self.vacancies_df.get('description', '')
                    ).apply(self.clean_text)
    
    def clean_text(self, text):
        """Очистка текста"""
        if not isinstance(text, str):
            return ""
        text = text.lower()
        text = re.sub(r'[^\w\s]', '', text)
        return text
    
    def init_models(self):
        """Инициализация моделей машинного обучения"""
        # LSTM модель для анализа текста
        self.lstm_model = None
        self.tokenizer = None
        
        try:
            if os.path.exists('lstm_model.h5') and os.path.exists('tokenizer_config.json'):
                self.lstm_model = load_model('lstm_model.h5')
                with open('tokenizer_config.json', 'r', encoding='utf-8') as f:
                    tokenizer_config = json.load(f)
                self.tokenizer = Tokenizer()
                self.tokenizer.from_json(tokenizer_config)
            else:
                self.train_lstm_model()
        except Exception as e:
            print(f"Ошибка при загрузке моделей: {e}")
            self.train_lstm_model()
        
        # # Item-Based Collaborative Filtering
        self.prepare_cf_model()
    
    def train_lstm_model(self):
        """Обучение LSTM модели на текстах вакансий"""
        texts = self.vacancies_df['combined_text'].tolist()
        
        # Токенизация текста
        self.tokenizer = Tokenizer(num_words=10000)
        self.tokenizer.fit_on_texts(texts)
        
        # Сохранение токенизатора
        tokenizer_json = self.tokenizer.to_json()
        with open('tokenizer_config.json', 'w', encoding='utf-8') as f:
            json.dump(tokenizer_json, f)
        
        # Преобразование текста в последовательности
        sequences = self.tokenizer.texts_to_sequences(texts)
        padded_sequences = pad_sequences(sequences, maxlen=200)
        
        # Создание и обучение модели
        self.lstm_model = Sequential([
            Embedding(input_dim=10000, output_dim=128, input_length=200),
            LSTM(128, dropout=0.2, recurrent_dropout=0.2),
            Dense(64, activation='relu'),
            Dropout(0.5),
            Dense(1, activation='sigmoid')
        ])
        
        self.lstm_model.compile(
            optimizer='adam',
            loss='binary_crossentropy',
            metrics=['accuracy']
        )
        
        # Здесь должна быть реальная разметка данных для обучения
        # В демонстрационных целях используем случайные метки
        dummy_labels = np.random.randint(0, 2, size=(len(texts),))
        
        self.lstm_model.fit(
            padded_sequences,
            dummy_labels,
            batch_size=32,
            epochs=5,
            validation_split=0.2,
            callbacks=[EarlyStopping(patience=2)]
        )
        
        self.lstm_model.save('lstm_model.h5')
    
    def prepare_cf_model(self):
        """Подготовка модели Item-Based Collaborative Filtering"""
        # Создаем матрицу признаков для вакансий
        features = pd.DataFrame()
        
        # Преобразуем категориальные признаки
        features['experience'] = self.vacancies_df['experience.name'].astype('category').cat.codes
        features['employment'] = self.vacancies_df['schedule.name'].astype('category').cat.codes
        
        # Обработка формата работы
        if 'work_format' in self.vacancies_df.columns:
            features['work_format'] = self.vacancies_df['work_format'].apply(
                lambda x: x[0]['name'] if isinstance(x, list) and len(x) > 0 else 'Не важно'
            ).astype('category').cat.codes
        else:
            features['work_format'] = 0  # Значение по умолчанию
        
        # Нормализуем числовые признаки
        features['salary_from'] = (self.vacancies_df['salary.from'] - self.vacancies_df['salary.from'].min()) / \
                                 (self.vacancies_df['salary.from'].max() - self.vacancies_df['salary.from'].min())
        features['salary_to'] = (self.vacancies_df['salary.to'] - self.vacancies_df['salary.to'].min()) / \
                               (self.vacancies_df['salary.to'].max() - self.vacancies_df['salary.to'].min())
        
        # Обработка текстовых признаков с помощью TF-IDF
        tfidf = TfidfVectorizer(max_features=100)
        tfidf_features = tfidf.fit_transform(self.vacancies_df['combined_text'])
        tfidf_df = pd.DataFrame(tfidf_features.toarray(), columns=tfidf.get_feature_names_out())
        
        # Объединяем все признаки
        features = pd.concat([features, tfidf_df], axis=1).fillna(0)
        
        # Создаем модель k-ближайших соседей
        self.cf_model = NearestNeighbors(n_neighbors=5, metric='cosine')
        self.cf_model.fit(features)
    
    def create_widgets(self):
        """Создание интерфейса пользователя"""
        main_frame = ttk.Frame(self.root, padding="10")
        main_frame.pack(fill=tk.BOTH, expand=True)

        # Блок названия должности
        ttk.Label(main_frame, text="Название текущей/желаемой должности:").grid(row=0, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.job_title, width=50).grid(row=0, column=1, sticky=tk.W, pady=5)

        # Блок резюме
        resume_frame = ttk.LabelFrame(main_frame, text="Резюме", padding=10)
        resume_frame.grid(row=1, column=0, columnspan=2, sticky=tk.EW, pady=10)

        ttk.Button(resume_frame, text="Ввести резюме", command=self.open_resume_editor).pack(side=tk.LEFT, padx=5)
        ttk.Button(resume_frame, text="Загрузить из файла", command=self.load_resume_file).pack(side=tk.LEFT, padx=5)

        # Блок опыта работы
        ttk.Label(main_frame, text="Опыт работы:").grid(row=2, column=0, sticky=tk.W, pady=5)
        experience_options = ["Нет опыта", "От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"]
        ttk.Combobox(main_frame, textvariable=self.experience, values=experience_options, state="readonly").grid(
            row=2, column=1, sticky=tk.W, pady=5)

        # Блок города
        ttk.Label(main_frame, text="Город:").grid(row=3, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.city, width=30).grid(row=3, column=1, sticky=tk.W, pady=5)

        # Чекбокс поиска в той же сфере
        ttk.Checkbutton(main_frame, text="Искать в той же сфере деятельности", variable=self.same_field,
                       command=self.toggle_job_field).grid(row=4, column=0, columnspan=2, sticky=tk.W, pady=5)

        # Блок типа занятости
        ttk.Label(main_frame, text="Тип занятости:").grid(row=5, column=0, sticky=tk.W, pady=5)
        employment_options = ["Полный день", "Удаленная работа", "Не важно"]
        ttk.Combobox(main_frame, textvariable=self.employment_type, values=employment_options, state="readonly", width=20).grid(row=5, column=1, sticky=tk.W, pady=5)

        # Блок формата работы
        ttk.Label(main_frame, text="Формат работы:").grid(row=6, column=0, sticky=tk.W, pady=5)
        format_options = ["На месте работодателя", "Удалённо", "Гибрид", "Не важно"]
        ttk.Combobox(main_frame, textvariable=self.work_format, values=format_options, state="readonly", 
                    width=20).grid(row=6, column=1, sticky=tk.W, pady=5)

        # Блок желаемой зарплаты
        ttk.Label(main_frame, text="Желаемая зарплата (руб.):").grid(row=7, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.desired_salary, width=15).grid(row=7, column=1, sticky=tk.W, pady=5)

        # Чекбокс безработного статуса
        ttk.Checkbutton(main_frame, text="Я в настоящее время безработный", variable=self.unemployed).grid(
            row=8, column=0, columnspan=2, sticky=tk.W, pady=5)

        # Кнопка поиска
        ttk.Button(main_frame, text="Найти вакансии", command=self.search_jobs).grid(
            row=9, column=0, columnspan=2, pady=20)

        # Кнопка выхода
        ttk.Button(main_frame, text="Выход", command=self.root.quit).grid(
            row=10, column=0, columnspan=2, pady=10)
    
    def toggle_job_field(self):
        """Обработчик изменения чекбокса 'Искать в той же сфере'"""
        if not self.same_field.get():
            self.open_desired_job_dialog()
    
    def open_desired_job_dialog(self):
        """Открытиетие диалога для указания желаемой должности"""
        dialog = tk.Toplevel(self.root)
        dialog.title("Укажите желаемую должность")
        dialog.geometry("400x200")

        ttk.Label(dialog, text="Введите желаемую должность:").pack(pady=10)
        ttk.Entry(dialog, textvariable=self.desired_job, width=40).pack(pady=5)

        ttk.Button(dialog, text="Сохранить", command=dialog.destroy).pack(pady=15)
    
    def open_resume_editor(self):
        """Открытие редактора резюме"""
        editor = tk.Toplevel(self.root)
        editor.title("Редактор резюме")
        editor.geometry("600x400")

        text_area = ScrolledText(editor, wrap=tk.WORD, width=70, height=20)
        text_area.pack(fill=ttk.BOTH, expand=True, padx=10, pady=10)

        if self.resume_text:
            text_area.insert(tk.END, self.resume_text)

        def save_resume():
            self.resume_text = text_area.get("1.0", tk.END)
            editor.destroy()
            messagebox.showinfo("Сохранено", "Резюме успешно сохранено")

        ttk.Button(editor, text="Сохранить", command=save_resume).pack(pady=5)
    
    def load_resume_file(self):
        """Загрузка резюме из файла"""
        file_path = filedialog.askopenfilename(
            filetypes=[("Текстовые файлы", "*.txt"), ("Документы Word", "*.docx"), ("PDF", "*.pdf")])
        if file_path:
            try:
                if file_path.endswith('.txt'):
                    with open(file_path, 'r', encoding='utf-8') as file:
                        self.resume_text = file.read()
                else:
                    self.resume_text = f"Файл {file_path} загружен (необходима обработка для извлечения текста)"
                messagebox.showinfo("Успех", "Резюме успешно загружено")
            except Exception as e:
                messagebox.showerror("Ошибка", f"Не удалось загрузить файл:\n{str(e)}")
    
    def search_jobs(self):
        """Поиск вакансий"""
        # Проверка обязательных полей
        if not self.job_title.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите название должности")
            return

        if not self.resume_text:
            messagebox.showerror("Ошибка", "Пожалуйста, введите или загрузите резюме")
            return

        if not self.city.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите город")
            return

        if not self.desired_salary.get() or not self.desired_salary.get().isdigit():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите валидную желаемую зарплату")
            return

        # Подготовка данных для поиска
        search_data = {
            "job_title": self.job_title.get(),
            "resume_text": self.resume_text,
            "experience": self.experience.get(),
            "city": self.city.get(),
            "same_field": self.same_field.get(),
            "desired_job": self.desired_job.get() if not self.same_field.get() else "",
            "employment_type": self.employment_type.get(),
            "work_format": self.work_format.get(),
            "desired_salary": int(self.desired_salary.get()),
            "unemployed": self.unemployed.get()
        }

        # Выбор метода поиска
        if self.same_field.get():
            # Используем LSTM для анализа текста резюме и вакансий
            filtered_jobs = self.filter_with_lstm(search_data)
        else:
            # Используем Item-Based Collaborative Filtering
            filtered_jobs = self.filter_with_cf(search_data)
        
        # Отображение результатов
        self.display_results(filtered_jobs)
    
    def filter_with_lstm(self, search_data):
        """Фильтрация вакансий с использованием LSTM модели"""
        # Базовые фильтры
        filtered_df = self.vacancies_df.copy()
        
        # Фильтр по городу
        if search_data['city']:
            filtered_df = filtered_df[filtered_df['area.name_desc'].str.contains(
                search_data['city'], case=False, na=False)]
        
        # Фильтр по опыту работы
        experience_map = {
            "Нет опыта": 0,
            "От 1 года до 3 лет": 1,
            "От 3 до 6 лет": 2,
            "Более 6 лет": 3
        }
        if search_data['experience'] in experience_map:
            filtered_df = filtered_df[filtered_df['experience.name'].apply(
    lambda x: experience_map[x] >= experience_map[search_data['experience']]
)]
        
        # Фильтр по типу занятости
        if search_data['employment_type'] != "Не важно":
            filtered_df = filtered_df[filtered_df['schedule.name'].str.contains(
                search_data['employment_type'], case=False, na=False)]
    
        
        # Фильтр по желаемой зарплате
        desired_salary = search_data['desired_salary']
        filtered_df = filtered_df[
            (filtered_df['salary.from'] <= desired_salary) | 
            (filtered_df['salary.to'] >= desired_salary)
        ]
        
        # Фильтр по формату работы
        if search_data['work_format'] != "Не важно":
            work_format_map = {
                "Удалённо": "REMOTE",
                "На месте работодателя": "ON_SITE",
                "Гибрид": "HYBRID"
            }
            format_key = work_format_map.get(search_data['work_format'], "")
            if format_key:
                filtered_df = filtered_df[filtered_df['work_format'].apply(
                    lambda x: any(f['id'].lower() == format_key.lower() for f in x) if isinstance(x, list) else False)]
        
        # Если после базовых фильтров осталось слишком много вакансий, применяем LSTM
        if len(filtered_df) > 10 and self.lstm_model is not None and self.tokenizer is not None:
            try:
                # Токенизация резюме и вакансий
                resume_seq = self.tokenizer.texts_to_sequences([search_data['resume_text']])
                resume_padded = pad_sequences(resume_seq, maxlen=200)
                
                job_seqs = self.tokenizer.texts_to_sequences(filtered_df['combined_text'])
                job_padded = pad_sequences(job_seqs, maxlen=600)
                
                # Получение эмбеддингов из LSTM
                lstm_layer = self.lstm_model.layers[-3]  # Предпоследний dense слой
                embedding_model = tf.keras.Model(
                    inputs=self.lstm_model.input,
                    outputs=lstm_layer.output
                )
                
                resume_embedding = embedding_model.predict(resume_padded)
                job_embeddings = embedding_model.predict(job_padded)
                
                # Вычисление косинусного сходства
                similarities = cosine_similarity(resume_embedding, job_embeddings)[0]
                filtered_df['similarity'] = similarities
                
                # Сортировка по сходству
                filtered_df = filtered_df.sort_values('similarity', ascending=False).head(10)
            except Exception as e:
                print(f"Ошибка при обработке LSTM: {e}")
        
        return filtered_df
    
    def filter_with_cf(self, search_data):
        """Фильтрация вакансий с использованием Item-Based Collaborative Filtering"""
        # Создаем профиль пользователя на основе введенных данных
        user_profile = pd.DataFrame()
        
        # Опыт работы
        experience_map = {
            "Нет опыта": 0,
            "От 1 года до 3 лет": 1,
            "От 3 до 6 лет": 2,
            "Более 6 лет": 3
        }
        user_profile['experience'] = [experience_map.get(search_data['experience'], 0)]
        
        # Тип занятости
        employment_map = {
            "Полный день": 0,
            "Удаленная работа": 1,
            "Не важно": 2
        }
        user_profile['employment'] = [employment_map.get(search_data['employment_type'], 2)]
        
        # Формат работы
        work_format_map = {
            "На месте работодателя": 0,
            "Удалённо": 1,
            "Гибрид": 2,
            "Не важно": 3
        }
        user_profile['work_format'] = [work_format_map.get(search_data['work_format'], 3)]
        
        # Желаемая зарплата (нормализованная)
        min_salary = self.vacancies_df['salary.from'].min()
        max_salary = self.vacancies_df['salary.from'].max()
        user_profile['salary_from'] = [
            (search_data['desired_salary'] - min_salary) / (max_salary - min_salary) if max_salary > min_salary else 0.5
        ]
        user_profile['salary_to'] = user_profile['salary_from']
        
        # Обработка текста желаемой должности с помощью TF-IDF
        tfidf = TfidfVectorizer(max_features=100)
        tfidf.fit(self.vacancies_df['combined_text'])
        
        desired_job_text = search_data['desired_job'] if search_data['desired_job'] else search_data['job_title']
        tfidf_features = tfidf.transform([desired_job_text])
        tfidf_df = pd.DataFrame(tfidf_features.toarray(), columns=tfidf.get_feature_names_out())
        
        # Объединяем все признаки
        user_profile = pd.concat([user_profile, tfidf_df], axis=1).fillna(0)
        
        # Находим ближайшие вакансии
        distances, indices = self.cf_model.kneighbors(user_profile)
        
        # Получаем отфильтрованные вакансии
        filtered_df = self.vacancies_df.iloc[indices[0]].copy()
        
        # Дополнительная фильтрация по городу
        if search_data['city']:
            filtered_df = filtered_df[filtered_df['area.name_desc'].str.contains(
                search_data['city'], case=False, na=False)]
        
        return filtered_df
    
    def display_results(self, filtered_jobs):
        """Отображение результатов поиска"""
        if filtered_jobs.empty:
            messagebox.showinfo("Результаты поиска", "К сожалению, не найдено подходящих вакансий.")
        else:
            # Создаем окно с результатами
            results_window = tk.Toplevel(self.root)
            results_window.title("Результаты поиска вакансий")
            results_window.geometry("800x600")
            
            # Создаем Treeview для отображения таблицы
            tree = ttk.Treeview(results_window, columns=('name', 'salary', 'company', 'experience', 'format','url'), show='headings')
            tree.heading('name', text='Должность')
            tree.heading('salary', text='Зарплата')
            tree.heading('company', text='Компания')
            tree.heading('experience', text='Опыт')
            tree.heading('format', text='Формат работы')
            tree.heading('url', text='Ссылка на вакансию')
            tree.column('name', width=200)
            tree.column('salary', width=100)
            tree.column('company', width=150)
            tree.column('experience', width=100)
            tree.column('format', width=150)
            tree.column('format', width=150)
            
            # Добавляем данные
            for _, row in filtered_jobs.iterrows():
                salary_from = row['salary.from']
                salary_to = row['salary.to']
                salary_str = f"{int(salary_from)}-{int(salary_to)}" if not pd.isna(salary_to) else str(int(salary_from))
                
                work_format = row.get('work_format', [])
                if isinstance(work_format, list) and len(work_format) > 0:
                    format_str = work_format[0].get('name', 'Не указано')
                else:
                    format_str = "Не указано"
                
                tree.insert('', tk.END, values=(
                    row.get('name_desc', 'Не указано'),
                    salary_str,
                    row.get('name_comp', 'Не указано'),
                    row.get('experience.name', 'Не указано'),
                    format_str
                ))
            
            tree.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)
            
            # Кнопка закрытия
            ttk.Button(results_window, text="Закрыть", command=results_window.destroy).pack(pady=10)

if __name__ == "__main__":
    df_vacancies = df_vacancies
    
    root = tk.Tk()
    app = JobSearchSystem(root, df_vacancies)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, filedialog
from tkinter.scrolledtext import ScrolledText
import pandas as pd
import numpy as np
import re
import os
import json
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import tensorflow as tf
from tensorflow.keras.utils import to_categorical


class JobSearchSystem:
    def __init__(self, root, vacancies_df):
        self.root = root
        self.root.title("Система поиска вакансий")
        self.root.geometry("900x650")
        
        # Загрузка данных
        self.vacancies_df = vacancies_df
        self.preprocess_data()
        
        # Инициализация моделей
        self.init_models()
        
        # Переменные интерфейса
        self.job_title = tk.StringVar()
        self.resume_text = ""
        self.experience = tk.StringVar(value="Нет опыта")
        self.city = tk.StringVar()
        self.unemployed = tk.BooleanVar(value=False)
        self.same_field = tk.BooleanVar(value=True)
        self.desired_job = tk.StringVar()
        self.desired_salary = tk.StringVar()
        self.employment_type = tk.StringVar(value="Полный день")
        self.work_format = tk.StringVar(value="Не важно")
        
        # Создание интерфейса
        self.create_widgets()
    
     def preprocess_data(self):
        """Предварительная обработка данных"""
        # Заполнение пропущенных значений
        self.vacancies_df['salary.from'] = self.vacancies_df['salary.from'].fillna(0)
        self.vacancies_df['salary.to'] = self.vacancies_df['salary.to'].fillna(0)
        self.vacancies_df['experience.name'] = self.vacancies_df['experience.name'].fillna('Нет опыта')
        
        # Очистка текстовых полей
        text_columns = ['name', 'description']
        for col in text_columns:
            if col in self.vacancies_df.columns:
                self.vacancies_df[col] = self.vacancies_df[col].fillna('').apply(self.clean_text)
        
        # Создаем комбинированный текст для анализа
            self.vacancies_df['combined_text'] = (
            self.vacancies_df.get('name', '') + " " + 
            self.vacancies_df.get('description', '')
        ).apply(self.clean_text)
        
        # Нормализация зарплат
        self.vacancies_df['salary_mid'] = (self.vacancies_df['salary.from'] + self.vacancies_df['salary.to']) / 2
    
    def clean_text(self, text):
        """Очистка текста"""
        if not isinstance(text, str):
            return ""
        text = text.lower()
        text = re.sub(r'[^\w\s]', '', text)
        return text
    
    def init_models(self):
        """Инициализация моделей машинного обучения"""
        self.lstm_model = None

        self.tokenizer = Tokenizer(num_words=10000)
        
        # Обучаем токенизатор на всех текстах
        texts = self.vacancies_df['combined_text'].tolist()
        self.tokenizer.fit_on_texts(texts) 
        
        # Пытаемся загрузить предобученную модель
        try:
            if os.path.exists('lstm_model.h5'):
                self.lstm_model = load_model('lstm_model.h5')
            else:
                self.train_lstm_model()
        except Exception as e:
            print(f"Ошибка при загрузке модели: {e}")
            self.train_lstm_model()
        
        # Подготовка модели рекомендаций
        self.prepare_recommendation_model()
    
    def train_lstm_model(self):
        """Обучение LSTM модели на текстах вакансий"""
        texts = self.vacancies_df['combined_text'].tolist()
        
        # Преобразование текста в последовательности
        sequences = self.tokenizer.texts_to_sequences(texts)
        padded_sequences = pad_sequences(sequences, maxlen=200)
        
        # Создаем искусственные метки для демонстрации
        num_classes = 5  # Упрощенная классификация по 5 категориям
        labels = np.random.randint(0, num_classes, size=(len(texts),))
        categorical_labels = to_categorical(labels, num_classes=num_classes)
        
        # Создание и обучение модели
        self.lstm_model = Sequential([
            Embedding(input_dim=10000, output_dim=128, input_length=200),
            LSTM(128, dropout=0.2, recurrent_dropout=0.2),
            Dense(64, activation='relu'),
            Dropout(0.5),
            Dense(num_classes, activation='softmax')
        ])
        
        self.lstm_model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        self.lstm_model.fit(
            padded_sequences,
            categorical_labels,
            batch_size=32,
            epochs=5,
            validation_split=0.2,
            callbacks=[EarlyStopping(patience=2)])
        
        self.lstm_model.save('lstm_model.h5')
    
    def prepare_recommendation_model(self):
        """Подготовка модели рекомендаций на основе TF-IDF и косинусного сходства"""
        # Создаем TF-IDF матрицу для текстов вакансий
        self.tfidf_vectorizer = TfidfVectorizer(max_features=500)
        self.tfidf_matrix = self.tfidf_vectorizer.fit_transform(self.vacancies_df['combined_text'])
    
    def create_widgets(self):
        """Создание интерфейса пользователя"""
        main_frame = ttk.Frame(self.root, padding="10")
        main_frame.pack(fill=tk.BOTH, expand=True)

        # Блок названия должности

        ttk.Label(main_frame, text="Название текущей/желаемой должности:").grid(row=0, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.job_title, width=50).grid(row=0, column=1, sticky=tk.W, pady=5)

        # Блок резюме
        resume_frame = ttk.LabelFrame(main_frame, text="Резюме", padding=10)
        resume_frame.grid(row=1, column=0, columnspan=2, sticky=tk.EW, pady=10)

        ttk.Button(resume_frame, text="Ввести резюме", command=self.open_resume_editor).pack(side=tk.LEFT, padx=5)
        ttk.Button(resume_frame, text="Загрузить из файла", command=self.load_resume_file).pack(side=tk.LEFT, padx=5)

        # Блок опыта работы
        ttk.Label(main_frame, text="Опыт работы:").grid(row=2, column=0, sticky=tk.W, pady=5)
        experience_options = ["Нет опыта", "От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"]
        ttk.Combobox(main_frame, textvariable=self.experience, values=experience_options, state="readonly").grid(
            row=2, column=1, sticky=tk.W, pady=5)

        # Блок города
        ttk.Label(main_frame, text="Город:").grid(row=3, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.city, width=30).grid(row=3, column=1, sticky=tk.W, pady=5)

        # Чекбокс поиска в той же сфере
        ttk.Checkbutton(main_frame, text="Искать в той же сфере деятельности", variable=self.same_field,
                       command=self.toggle_job_field).grid(row=4, column=0, columnspan=2, sticky=tk.W, pady=5)

        # Блок типа занятости
        ttk.Label(main_frame, text="Тип занятости:").grid(row=5, column=0, sticky=tk.W, pady=5)
        employment_options = ["Полный день", "Удаленная работа", "Не важно"]
        ttk.Combobox(main_frame, textvariable=self.employment_type, values=employment_options, state="readonly", width=20).grid(row=5, column=1, sticky=tk.W, pady=5)

        # Блок формата работы
        ttk.Label(main_frame, text="Формат работы:").grid(row=6, column=0, sticky=tk.W, pady=5)
        format_options = ["На месте работодателя", "Удалённо", "Гибрид", "Не важно"]
        ttk.Combobox(main_frame, textvariable=self.work_format, values=format_options, state="readonly", 
                    width=20).grid(row=6, column=1, sticky=tk.W, pady=5)

        # Блок желаемой зарплаты
        ttk.Label(main_frame, text="Желаемая зарплата (руб.):").grid(row=7, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.desired_salary, width=15).grid(row=7, column=1, sticky=tk.W, pady=5)

        # Чекбокс безработного статуса
        ttk.Checkbutton(main_frame, text="Я в настоящее время безработный", variable=self.unemployed).grid(
            row=8, column=0, columnspan=2, sticky=tk.W, pady=5)

        # Кнопка поиска
        ttk.Button(main_frame, text="Найти вакансии", command=self.search_jobs).grid(
            row=9, column=0, columnspan=2, pady=20)

        # Кнопка выхода
        ttk.Button(main_frame, text="Выход", command=self.root.quit).grid(
            row=10, column=0, columnspan=2, pady=10)
    
    def toggle_job_field(self):
        """Обработчик изменения чекбокса 'Искать в той же сфере'"""
        if not self.same_field.get():
            self.open_desired_job_dialog()
    
    def open_desired_job_dialog(self):
        """Открытие диалога для указания желаемой должности"""
        dialog = tk.Toplevel(self.root)
        dialog.title("Укажите желаемую должность")
        dialog.geometry("400x200")

        ttk.Label(dialog, text="Введите желаемую должность:").pack(pady=10)
        ttk.Entry(dialog, textvariable=self.desired_job, width=40).pack(pady=5)

        ttk.Button(dialog, text="Сохранить", command=dialog.d.destroy).pack(pady=15)
    
    def open_resume_editor(self):
        """Открытие редактора резюме"""
        editor = tk.Toplevel(self.root)
        editor.title("Редактор резюме")
        editor.geometry("600x400")

        text_area = ScrolledText(editor, wrap=tk.WORD, width=70, height=20)
        text_area.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)

        if self.resume_text:
            text_area.insert(tk.END, self.resume_text)

        def save_resume():
            self.resume_text = text_area.get("1.0", tk.END)
            editor.destroy()
            messagebox.showinfo("Сохранено", "Резюме успешно сохранено")

        ttk.Button(editor, text="Сохранить", command=save_resume).pack(pady=5)
    
    def load_resume_file(self):
        """Загрузка резюме из файла"""
        file_path = filedialog.askopenfilename(
            filetypes=[("Текстовые файлы", "*.txt"), ("Документы Word", "*.docx"), ("PDF", "*.pdf")])
        if file_path:
            try:
                if file_path.endswith('.txt'):
                    with open(file_path, 'r', encoding='utf-8') as file:
                        self.resume_text = file.read()
                else:
                    self.resume_text = f"Файл {file_path} загружен (необходима обработка для извлечения текста)"
                messagebox.showinfo("Успех", "Резюме успешно загружено")
            except Exception as e:
                messagebox.showerror("Ошибка", f"Не удалось загрузить файл:\n{str(e)}")
    
    def search_jobs(self):
        """Поиск вакансий"""
        # Проверка обязательных пол полей
        if not self.job_title.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите название должности")
            return

        if not self.resume_text:
            messagebox.showerror("Ошибка", "Пожалуйста, введите или загрузите резюме")
            return

        if not self.city.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите город")
            return

        if not self.desired_salary.get() or not self.desired_salary.get().isdigit():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите валидную желаемую зарплату")
            return

        # Подготовка данных для поиска
        search_data = {
            "job_title": self.job_title.get(),
            "resume_text": self.resume_text,
            "experience": self.experience.get(),
            "city": self.city.get(),
            "same_field": self.same_field.get(),
            "desired_job": self.desired_job.get() if not self.same_field.get() else "",
            "employment_type": self.employment_type.get(),
            "work_format": self.work_format.get(),
            "desired_salary": int(self.desired_salary.get()),
            "unemployed": self.unemployed.get()
        }

        # Поиск вакансий
        filtered_jobs = self.find_relevant_jobs(search_data)
        
        # Отображение результатов
        self.display_results(filtered_jobs)
    
    def find_relevant_jobs(self, search_data):
        """Поиск релевантных вакансий"""
        # Базовые фильтры
        filtered_df = self.vacancies_df.copy()
        
        # Фильтр по городу
        if search_data['city']:
            filtered_df = filtered_df[filtered_df['processed_area.name_desc'].str.contains(
                search_data['city'], case=False, na=False)]
        
        # Фильтр по опыту работы
        experience_map = {
            "Нет опыта": ["Нет опыта"],
            "От 1 года до 3 лет": ["От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"],
            "От 3 до 6 лет": ["От 3 до 6 лет", "Более 6 лет"],
            "Более 6 лет": ["Более 6 лет"]
        }
        if search_data['experience'] in experience_map:
            filtered_df = filtered_df[filtered_df['experience.name'].isin(
                experience_map[search_data['experience']])]
        
        # Фильтр по типу занятости
        if search_data['employment_type'] != "Не важно":
            filtered_df = filtered_df[filtered_df['schedule.name'].str.contains(
                search_data['employment_type'], case=False, na=False)]
        
        # Фильтр по желаемой зарплате
        desired_salary = search_data['desired_salary']
        filtered_df = filtered_df[
            (filtered_df['salary.from'] <= desired_salary) | 
            (filtered_df['salary.to'] >= desired_salary)
        ]
        
        # Если после базовых фильтров осталось слишком много вакансий, применяем TF-IDF
        if len(filtered_df) > 0:
            # Получаем вектор резюме
            resume_vector = self.tfidf_vectorizer.transform([search_data['resume_text']])
            
            # Получаем векторы вакансий
            job_vectors = self.tfidf_vectorizer.transform(filtered_df['combined_text'])
            
            # Вычисляем косинусное сходство
            similarities = cosine_similarity(resume_vector, job_vectors)[0]
            filtered_df['similarity'] = similarities
            
            # Сортируем по сходству
            filtered_df = filtered_df.sort_values('similarity', ascending=False)
            
            # Ограничиваем количество результатов
            filtered_df = filtered_df.head(10)
        
        return filtered_df
    
    def display_results(self, filtered_jobs):
        """Отображение результатов поиска"""
        if filtered_jobs.empty:
            messagebox.showinfo("Результаты поиска", "К сожалению, не найдено подходящих вакансий.")
        else:
            # Создаем окно с результатами
            results_window = tk.Toplevel(self.root)
            results_window.title("Результаты поиска вакансий")
            results_window.geometry("800x600")
            
            # Создаем Treeview для отображения таблицы
            tree = ttk.Treeview(results_window, columns=('name', 'description', 'url', 'salary'), show='headings')
            tree.heading('name', text='Должность')
            tree.heading('description', text='Описание вакансии')
            tree.heading('url', text='Ссылка на вакансию')
            tree.heading('salary', text='Заработная плата')
            
            tree.column('name', width=200)
            tree.column('description', width=300)
            tree.column('url', width=150)
            tree.column('salary', width=100)
            
            # Добавляем данные
            for _, row in filtered_jobs.iterrows():
                salary_from = row['salary.from']
                salary_str = f"{int(salary_from)}" if not pd.isna(salary_from) else "Не указано"
                
                tree.insert('', tk.END, values=(
                    row.get('name_desc', 'Не указано'),
                    row.get('processed_description_desc', 'Не указано'),
                    row.get('vacancies_url', 'Не указано'),
                    salary_str
                ))
            
            tree.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)
            
            # Кнопка закрытия
            ttk.Button(results_window, text="Закрыть", command=results_window.destroy).pack(pady=10)


if __name__ == "__main__": 
    try:
        df_vacancies = pd.read_csv('vacancies.csv')
        root = tk.Tk()
        app = JobSearchSystem(root, df_vacancies)
        root.mainloop()
    except Exception as e:
        print(f"Ошибка при запуске приложения: {e}")
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, filedialog
from tkinter.scrolledtext import ScrolledText
import pandas as pd
import numpy as np
import re
import os
import json
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


class JobSearchSystem:
    def __init__(self, root, vacancies_df):
        self.root = root
        self.root.title("Система поиска вакансий")
        self.root.geometry("900x650")
        
        # Загрузка данных
        self.vacancies_df = vacancies_df
        self.preprocess_data()
        
        # Инициализация моделей
        self.init_models()
        
        # Переменные интерфейса
        self.job_title = tk.StringVar()
        self.resume_text = ""
        self.experience = tk.StringVar(value="Нет опыта")
        self.city = tk.StringVar()
        self.unemployed = tk.BooleanVar(value=False)
        self.same_field = tk.BooleanVar(value=True)
        self.desired_job = tk.StringVar()
        self.desired_salary = tk.StringVar()
        self.employment_type = tk.StringVar(value="Полный день")
        self.work_format = tk.StringVar(value="Не важно")
        
        # Создание интерфейса
        self.create_widgets()
    
    def preprocess_data(self):
        """Предварительная обработка данных"""
        # Заполнение пропущенных значений
        self.vacancies_df['salary.from'] = self.vacancies_df['salary.from'].fillna(0)
        self.vacancies_df['salary.to'] = self.vacancies_df['salary.to'].fillna(0)
        self.vacancies_df['experience.name'] = self.vacancies_df['experience.name'].fillna('Нет опыта')
        
        # Очистка текстовых полей
        text_columns = ['name', 'description']
        for col in text_columns:
            if col in self.vacancies_df.columns:
                self.vacancies_df[col] = self.vacancies_df[col].fillna('').apply(self.clean_text)
        
        # Создаем комбинированный текст для анализа
        self.vacancies_df['combined_text'] = (
            self.vacancies_df.get('name', '') + " " + 
            self.vacancies_df.get('description', '')
        ).apply(self.clean_text)
        
        # Нормализация зарплат
        self.vacancies_df['salary_mid'] = (self.vacancies_df['salary.from'] + self.vacancies_df['salary.to']) / 2
        
        # Подготовка данных для Item-Based CF
        self.prepare_item_based_data()
        
        # Подготовка данных для LSTM
        self.prepare_lstm_data()
    
    def prepare_lstm_data(self):
        """Подготовка данных для LSTM модели"""
        # Создаем категории на основе профессиональных областей (пример)
        if 'professional_roles' in self.vacancies_df.columns:
            # Если есть информация о профессиональных ролях, используем ее
            self.vacancies_df['category'] = self.vacancies_df['professional_roles'].apply(
                lambda x: x[0]['name'] if isinstance(x, list) and len(x) > 0 else 'Другое'
            )
        else:
            # Иначе создаем искусственные категории на основе названий должностей
            self.vacancies_df['category'] = self.vacancies_df['name'].apply(
                lambda x: self.detect_category(x)
            )
        
        # Кодируем категории
        self.label_encoder = LabelEncoder()
        self.vacancies_df['category_encoded'] = self.label_encoder.fit_transform(
            self.vacancies_df['category']
        )
        
        # Сохраняем количество классов
        self.num_classes = len(self.label_encoder.classes_)
    
    def detect_category(self, job_title):
        """Определение категории на основе названия должности"""
        job_title = job_title.lower()
        if any(word in job_title for word in ['разработчик', 'developer', 'программист', 'engineer']):
            return 'Разработка'
        elif any(word in job_title for word in ['аналитик', 'analyst', 'data scientist']):
            return 'Аналитика'
        elif any(word in job_title for word in ['менеджер', 'manager', 'руководитель']):
            return 'Менеджмент'
        elif any(word in job_title for word in ['дизайн', 'designer']):
            return 'Дизайн'
        else:
            return 'Другое'
    def prepare_item_based_data(self):
        """Подготовка данных для Item-Based Collaborative Filtering"""
        # Создаем матрицу признаков для вакансий
        self.job_features = pd.get_dummies(
            self.vacancies_df[['experience.name', 'schedule.name']], 
            columns=['experience.name', 'schedule.name'])

        # Добавляем нормализованную зарплату
        self.job_features['salary_norm'] = (
            self.vacancies_df['salary_mid'] - self.vacancies_df['salary_mid'].min()
        ) / (self.vacancies_df['salary_mid'].max() - self.vacancies_df['salary_mid'].min())

        # Заполняем пропуски
        self.job_features = self.job_features.fillna(0)

        # Создаем словарь для быстрого поиска индексов по названию должности
        # Используем очищенные названия должностей
        self.job_title_to_index = {
            self.clean_text(title): idx 
            for idx, title in enumerate(self.vacancies_df['name'])
        }

        # Создаем матрицу сходства для Item-Based CF
        self.item_similarity = 1 - pairwise_distances(
            self.job_features, 
            metric='cosine')
  

    def clean_text(self, text):
        """Очистка текста"""
        if not isinstance(text, str):
            return ""
        text = text.lower()
        text = re.sub(r'[^\w\s]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        return text
    
    def init_models(self):
        """Инициализация моделей машинного обучения"""
        self.lstm_model = None
        self.tokenizer = Tokenizer(num_words=10000)
        
        # Обучаем токенизатор на всех текстах
        texts = self.vacancies_df['combined_text'].tolist()
        self.tokenizer.fit_on_texts(texts) 
        
        # Пытаемся загрузить предобученную модель
        try:
            if os.path.exists('lstm_model.h5'):
                self.lstm_model = load_model('lstm_model.h5')
                # Загружаем историю обучения для отображения графиков
                if os.path.exists('training_history.json'):
                    with open('training_history.json', 'r') as f:
                        self.training_history = json.load(f)
            else:
                self.train_lstm_model()
        except Exception as e:
            print(f"Ошибка при загрузке модели: {e}")
            self.train_lstm_model()
        
        # Подготовка модели рекомендаций
        self.prepare_recommendation_model()
    
    def train_lstm_model(self):
        """Обучение LSTM модели на текстах вакансий"""
        texts = self.vacancies_df['combined_text'].tolist()
        labels = self.vacancies_df['category_encoded'].values
        
        # Преобразование текста в последовательности
        sequences = self.tokenizer.texts_to_sequences(texts)
        padded_sequences = pad_sequences(sequences, maxlen=200)
        
        # Разделение на обучающую и тестовую выборки
        X_train, X_test, y_train, y_test = train_test_split(
            padded_sequences, labels, test_size=0.2, random_state=42
        )
        
        # Преобразование меток в one-hot encoding
        y_train = to_categorical(y_train, num_classes=self.num_classes)
        y_test = to_categorical(y_test, num_classes=self.num_classes)
        
        # Создание модели
        self.lstm_model = Sequential([
            Embedding(input_dim=10000, output_dim=128, input_length=200),
            LSTM(128, dropout=0.2, recurrent_dropout=0.2, return_sequences=True),
            LSTM(64, dropout=0.2, recurrent_dropout=0.2),
            Dense(64, activation='relu'),
            Dropout(0.5),
            Dense(self.num_classes, activation='softmax')
        ])
        
        self.lstm_model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        # Коллбэки
        callbacks = [
            EarlyStopping(patience=3, restore_best_weights=True),
            ModelCheckpoint('lstm_model.h5', save_best_only=True)
        ]
        
        # Обучение модели
        history = self.lstm_model.fit(
            X_train,
            y_train,
            batch_size=64,
            epochs=20,
            validation_data=(X_test, y_test),
            callbacks=callbacks
        )
        
        # Сохраняем историю обучения
        self.training_history = history.history
        with open('training_history.json', 'w') as f:
            json.dump(self.training_history, f)
        
        # Визуализация обучения
        self.plot_training_history()
    
    def plot_training_history(self):
        """Визуализация процесса обучения модели"""
        plt.figure(figsize=(12, 5))
        
        # График точности
        plt.subplot(1, 2, 1)
        plt.plot(self.training_history['accuracy'], label='Точность на обучении')
        plt.plot(self.training_history['val_accuracy'], label='Точность на валидации')
        plt.title('Точность модели')
        plt.ylabel('Точность')
        plt.xlabel('Эпоха')
        plt.legend()
        
        # График потерь
        plt.subplot(1, 2, 2)
        plt.plot(self.training_history['loss'], label='Потери на обучении')
        plt.plot(self.training_history['val_loss'], label='Потери на валидации')
        plt.title('Потери модели')
        plt.ylabel('Потери')
        plt.xlabel('Эпоха')
        plt.legend()
        
        plt.tight_layout()
        plt.savefig('training_history.png.png')
        plt.close()
    
    def prepare_recommendation_model(self):
        """Подготовка модели рекомендаций на основе TF-IDF и косинусного сходства"""
        # Создаем TF-IDF матрицу для текстов вакансий
        self.tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
        self.tfidf_matrix = self.tfidf_vectorizer.fit_transform(self.vacancies_df['combined_text'])
        
        # Создаем матрицу сходства для Item-Based CF
        self.item_similarity = 1 - pairwise_distances(
            self.job_features, 
            metric='cosine'
        )
        
        # Модель KNN для рекомендаций
        self.knn_model = NearestNeighbors(n_neighbors=10, metric='cosine')
        self.knn_model.fit(self.tfidf_matrix)
    
    def create_widgets(self):
        """Создание интерфейса пользователя"""
        main_frame = ttk.Frame(self.root, padding="10")
        main_frame.pack(fill=tk.BOTH, expand=True)

        # Блок названия должности
        ttk.Label(main_frame, text="Название текущей/желаемой должности:").grid(row=0, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.job_title, width=50).grid(row=0, column=1, sticky=tk.W, pady=5)

        # Блок резюме
        resume_frame = ttk.LabelFrame(main_frame, text="Резюме", padding=10)
        resume_frame.grid(row=1, column=0, columnspan=2, sticky=tk.EW, pady=10)

        ttk.Button(resume_frame, text="Ввести резюме", command=self.open_resume_editor).pack(side=tk.LEFT, padx=5)
        ttk.Button(resume_frame, text="Загрузить из файла", command=self.load_resume_file).pack(side=tk.LEFT, padx=5)

        # Блок опыта работы
        ttk.Label(main_frame, text="Опыт работы:").grid(row=2, column=0, sticky=tk.W, pady=5)
        experience_options = ["Нет опыта", "От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"]
        ttk.Combobox(main_frame, textvariable=self.experience, values=experience_options, state="readonly").grid(
            row=2, column=1, sticky=tk.W, pady=5)

        # Блок города
        ttk.Label(main_frame, text="Город:").grid(row=3, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.city, width=30).grid(row=3, column=1, sticky=tk.W, pady=5)

        # Чекбокс поиска в той же сфере
        ttk.Checkbutton(main_frame, text="Искать в той же сфере деятельности", variable=self.same_field,
                       command=self.toggle_job_field).grid(row=4, column=0, columnspan=2, sticky=tk.W, pady=5)

        # Блок типа занятости
        ttk.Label(main_frame, text="Тип занятости:").grid(row=5, column=0, sticky=tk.W, pady=5)
        employment_options = ["Полный день", "Удаленная работа", "Не важно"]
        ttk.Combobox(main_frame, textvariable=self.employment_type, values=employment_options, state="readonly", width=20).grid(row=5, column=1, sticky=tk.W, pady=5)

        # Блок формата работы
        ttk.Label(main_frame, text="Формат работы:").grid(row=6, column=0, sticky=tk.W, pady=5)
        format_options = ["На месте работодателя", "Удалённо", "Гибрид", "Не важно"]
        ttk.Combobox(main_frame, textvariable=self.work_format, values=format_options, state="readonly", 
                    width=20).grid(row=6, column=1, sticky=tk.W, pady=5)

        # Блок желаемой зарплаты
        ttk.Label(main_frame, text="Желаемая зарплата (руб.):").grid(row=7, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.desired_salary, width=15).grid(row=7, column=1, sticky=tk.W, pady=5)

        # Чекбокс безработного статуса
        ttk.Checkbutton(main_frame, text="Я в настоящее время безработный", variable=self.unemployed).grid(
            row=8, column=0, columnspan=2, sticky=tk.W, pady=5)

        # Кнопка поиска
        ttk.Button(main_frame, text="Найти вакансии", command=self.search_jobs).grid(
            row=9, column=0, columnspan=2, pady=20)

        # Кнопка выхода
        ttk.Button(main_frame, text="Выход", command=self.root.quit).grid(
            row=10, column=0, columnspan=2, pady=10)
    
    def toggle_job_field(self):
        """Обработчик изменения чекбокса 'Искать в той же сфере'"""
        if not self.same_field.get():
            self.open_desired_job_dialog()
    
    def open_desired_job_dialog(self):
        """Открытие диалога для указания желаемой должности"""
        dialog = tk.Toplevel(self.root)
        dialog.title("Укажите желаемую должность")
        dialog.geometry("400x200")

        ttk.Label(dialog, text="Введите желаемую должность:").pack(pady=10)
        entry = ttk.Entry(dialog, textvariable=self.desired_job, width=40)
        entry.pack(pady=5)
        entry.focus_set()

        def save_and_close():
            if not self.desired_job.get():
                messagebox.showerror("Ошибка", "Пожалуйста, введите желаемую должность")
                return
            dialog.destroy()

        ttk.Button(dialog, text="Сохранить", command=save_and_close).pack(pady=15)
    
    def open_resume_editor(self):
        """Открытие редактора резюме"""
        editor = tk.Toplevel(self.root)
        editor.title("Редактор резюме")
        editor.geometry("600x400")

        text_area = ScrolledText(editor, wrap=tk.WORD, width=70, height=20)
        text_area.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)

        if self.resume_text:
            text_area.insert(tk.END, self.resume_text)

        def save_resume():
            self.resume_text = text_area.get("1.0", tk.END)
            editor.destroy()
            messagebox.showinfo("Сохранено", "Резюме успешно сохранено")

        ttk.Button(editor, text="Сохранить", command=save_resume).pack(pady=5)
    
    def load_resume_file(self):
        """Загрузка резюме из файла"""
        file_path = filedialog.askopenfilename(
            filetypes=[("Текстовые файлы", "*.txt"), ("Документы Word", "*.docx"), ("PDF", "*.pdf")])
        if file_path:
            try:
                if file_path.endswith('.txt'):
                    with open(file_path, 'r', encoding='utf-8') as file:
                        self.resume_text = file.read()
                else:
                    self.resume_text = f"Файл {file_path} загружен (необходима обработка для извлечения текста)"
                messagebox.showinfo("Успех", "Резюме успешно загружено")
            except Exception as e:
                messagebox.showerror("Ошибка", f"Не удалось загрузить файл:\n{str(e)}")
    
    def search_jobs(self):
        """Поиск вакансий"""
        # Проверка обязательных полей
        if not self.job_title.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите название должности")
            return

        if not self.resume_text:
            messagebox.showerror("Ошибка", "Пожалуйста, введите или загрузите резюме")
            return

        if not self.city.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите город")
            return

        if not self.desired_salary.get() or not self.desired_salary.get().isdigit():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите валидную желаемую зарплату")
            return

        # Проверка для случая, когда снята галочка "Искать в той же сфере"
        if not self.same_field.get() and not self.desired_job.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите желаемуюую должность")
            return

        # Подготовка данных для поиска
        search_data = {
            "job_title": self.job_title.get(),
            "resume_text": self.resume_text,
            "experience": self.experience.get(),
            "city": self.city.get(),
            "same_field": self.same_field.get(),
            "desired_job": self.desired_job.get() if not self.same_field.get() else "",
            "employment_type": self.employment_type.get(),
            "work_format": self.work_format.get(),
            "desired_salary": int(self.desired_salary.get()),
            "unemployed": self.unemployed.get()
        }

        # Поиск вакансий
        if self.same_field.get():
            filtered_jobs = self.find_relevant_jobs(search_data)
        else:
            filtered_jobs = self.find_jobs_with_item_based_cf(search_data)
        
        # Отображение результатов
        self.display_results(filtered_jobs)
    
    def find_relevant_jobs(self, search_data):
        """Поиск релевантных вакансий с использованием TF-IDF и LSTM"""
        # Базовые фильтры
        filtered_df = self.vacancies_df.copy()
        
        # Фильтр по городу
        if search_data['city']:
            filtered_df = filtered_df[filtered_df['area.name'].str.contains(
                search_data['city'], case=False, na=False)]
        
        # Фильтр по опыту работы
        experience_map = {
            "Нет опыта": ["Нет опыта"],
            "От 1 года до 3 лет": ["От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"],
            "От 3 до 6 лет": ["От 3 до 6 лет", "Более 6 лет"],
            "Более 6 лет": ["Более 6 лет"]
        }
        if search_data['experience'] in experience_map:
            filtered_df = filtered_df[filtered_df['experience.name'].isin(experience_map[search_data['experience']])]
        
        # Фильтр по типу занятости
        if search_data['employment_type'] != "Не важно":
            filtered_df = filtered_df[filtered_df['schedule.name'].str.contains(
                search_data['employment_type'], case=False, na=False)]
        
        # Фильтр по желаемой зарплате
        desired_salary = search_data['desired_salary']
        filtered_df = filtered_df[
            (filtered_df['salary.from'] <= desired_salary) | 
            (filtered_df['salary.to'] >= desired_salary)
        ]
        
        # Если после базовых фильтров осталось слишком много вакансий, применяем TF-IDF и LSTM
        if len(filtered_df) > 0:
            # Получаем вектор резюме для TF-IDF
            resume_vector = self.tfidf_vectorizer.transform([search_data['resume_text']])
            
            # Получаем векторы вакансий
            job_vectors = self.tfidf_vectorizer.transform(filtered_df['combined_text'])
            
            # Вычисляем косинусное сходство для TF TF-IDF
            similarities_tfidf = cosine_similarity(resume_vector, job_vectors)[0]
            
            # Получаем предсказания категории LSTM для резюме
            resume_sequence = self.tokenizer.texts_to_sequences([search_data['resume_text']])
            resume_padded = pad_sequences(resume_sequence, maxlen=200)
            resume_category_prob = self.lstm_model.predict(resume_padded)[0]
            
            # Получаем предсказания категорий для вакансий
            job_sequences = self.tokenizer.texts_to_sequences(filtered_df['combined_text'].tolist())
            job_padded = pad_sequences(job_sequences, maxlen=200)
            job_categories_prob = self.lstm_model.predict(job_padded)
            
            # Вычисляем сходство категорий (косинусное сходство между векторами вероятностей)
            similarities_lstm = cosine_similarity(
                resume_category_prob.reshape(1, -1), 
                job_categories_prob
            )[0]
            
            # Комбинируем оценки (можно настроить веса)
            combined_similarity = 0.6 * similarities_tfidf + 0.4 * similarities_lstm
            
            filtered_df['similarity'] = combined_similarity
            filtered_df = filtered_df.sort_values('similarity', ascending=False)
            
            # Ограничиваем количество результатов
            filtered_df = filtered_df.head(10)
        
        return filtered_df
    def find_jobs_with_item_based_cf(self, search_data):
        current_job_title = search_data['job_title'].lower()
        current_job_idx = self.job_title_to_index.get(current_job_title, None)
    
        if current_job_idx is None:
        # Если текущая должность не найдена, ищем похожие по названию
            similar_titles = [title for title in self.job_title_to_index.keys() 
                        if current_job_title in title]
        
        if not similar_titles:
            messagebox.showwarning("Предупреждение", 
                                f"Должность '{search_data['job_title']}' не найдена в базе. Используется стандартный поиск.")
            return self.find_relevant_jobs(search_data)
        
        # Берем первую похожую должность
        current_job_idx = self.job_title_to_index[similar_titles[0]]
    
    # Получаем рекомендации на основе сходства
        similar_indices = np.argsort(-self.item_similarity[current_job_idx])[1:11]  # Берем топ-10 похожих
    
    # Фильтруем по другим параметрам (город, зарплата и т.д.)
        filtered_jobs = self.vacancies_df.iloc[similar_indices].copy()
    
    # Фильтр по городу
        if search_data['city']:
            filtered_jobs = filtered_jobs[
            filtered_jobs['area.name'].str.contains(search_data['city'], case=False, na=False)]
    
    # Фильтр по опыту работы
        experience_map = {
        "Нет опыта": ["Нет опыта"],
        "От 1 года до 3 лет": ["От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"],
        "От 3 до 6 лет": ["От 3 до 6 лет", "Более 6 лет"],
        "Более 6 лет": ["Более 6 лет"]
    }
        if search_data['experience'] in experience_map:
            filtered_jobs = filtered_jobs[
            filtered_jobs['experience.name'].isin(experience_map[search_data['experience']])
        ]
    
    # Фильтр по типу занятости (исправлена опечатка "НеНе важно")
        if search_data['employment_type'] != "Не важно":
            filtered_jobs = filtered_jobs[
            filtered_jobs['schedule.name'].str.contains(
                search_data['employment_type'], case=False, na=False)
        ]
    
    # Фильтр по желаемой зарплате
        desired_salary = search_data['desired_salary']
        filtered_jobs = filtered_jobs[
        (filtered_jobs['salary.from'] <= desired_salary) | 
        (filtered_jobs['salary.to'] >= desired_salary)
    ]
    
    # Добавляем оценку сходства
        filtered_jobs['similarity'] = self.item_similarity[current_job_idx][similar_indices]
        filtered_jobs = filtered_jobs.sort_values('similarity', ascending=False)
    
        return filtered_jobs.head(10)

    
   
    
    def display_results(self, filtered_jobs):
        """Отображение результатов поиска"""
        if filtered_jobs.empty:
            messagebox.showinfo("Результаты поиска", "К сожалению, не найдено подходящих вакансий.")
        else:
            # Создаем окно с результатами
            results_window = tk.Toplevel(self.root)
            results_window.title("Результаты поиска вакансий")
            results_window.geometry("900x600")
            
            # Создаем Treeview для отображения таблицы
            tree = ttk.Treeview(results_window, columns=('name', 'company', 'salary', 'experience', 'similarity'), show='headings')
            tree.heading('name', text='Должность')
            tree.heading('company', text='Компания')
            tree.heading('salary', text='Зарплата')
            tree.heading('experience', text='Опыт')
            tree.heading('similarity', text='Сходство')
            
            tree.column('name', width=200)
            tree.column('company', width=150)
            tree.column('salary', width=100)
            tree.column('experience', width=100)
            tree.column('similarity', width=80)
            
            # Добавляем данные
            for _, row in filtered_jobs.iterrows():
                salary_from = row['salary.from']
                salary_to = row['salary.to']
                
                if pd.notna(salary_from) and pd.notna(salary_to):
                    salary_str = f"{int(salary_from)}-{int(salary_to)}"
                elif pd.notna(salary_from):
                    salary_str = f"от {int(salary_from)}"
                elif pd.notna(salary_to):
                    salary_str = f"до {int(salary_to)}"
                else:
                    salary_str = "Не указано"
                
                similarity = row.get('similarity', 0)
                similarity_str = f"{similarity:.2f}" if isinstance(similarity, (float, np.float64)) else "N/A"
                
                tree.insert('', tk.END, values=(
                    row.get('name', 'Не указано'),
                    row.get('employer.name', 'Не указано'),
                    salary_str,
                    row.get('experience.name', 'Не указано'),
                    similarity_str
                ))
            
            # Добавляем полосу прокрутки
            scrollbar = ttk.Scrollbar(results_window, orient="vertical", command=tree.yview)
            tree.configure(yscrollcommand=scrollbar.set)
            scrollbar.pack(side="right", fill="y")
            tree.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)
            
            # Кнопка закрытия
            ttk.Button(results_window, text="Закрыть", command=results_window.destroy).pack(pady=10)


if __name__ == "__main__":
    # Загрузка реального датасета
    try:
        # Пытаемся загрузить данные из CSV файла
        df_vacancies = pd.read_csv('vacancies.csv', low_memory=False)
        
        # Если в данных есть JSON-столбцы, преобразуем их
        json_columns = ['professional_roles', 'employment', 'experience', 'salary']
        for col in json_columns:
            if col in df_vacancies.columns:
                df_vacancies[col] = df_vacancies[col].apply(
                    lambda x: json.loads(x.replace("'", "\"")) if isinstance(x, str) else x
                )
        
        # Извлекаем вложенные данные
        if 'professional_roles' in df_vacancies.columns:
            df_vacancies['professional_roles'] = df_vacancies['professional_roles'].apply(
                lambda x: x[0] if isinstance(x, list) and len(x) > 0 else {}
            )
            df_vacancies['professional_role_name'] = df_vacancies['professional_roles'].apply(
                lambda x: x.get('name', '') if isinstance(x, dict) else ''
            )
        
        if 'experience' in df_vacancies.columns:
            df_vacancies['experience.name'] = df_vacancies['experience'].apply(
                lambda x: x.get('name', 'Нет опыта') if isinstance(x, dict) else 'Нет опыта'
            )
        
        if 'employment' in df_vacancies.columns:
            df_vacancies['schedule.name'] = df_vacancies['employment'].apply(
                lambda x: x.get('name', '') if isinstance(x, dict) else ''
            )
        
        if 'salary' in df_vacancies.columns:
            df_vacancies['salary.from'] = df_vacancies['salary'].apply(
                lambda x: x.get('from', None) if isinstance(x, dict) else None
            )
            df_vacancies['salary.to'] = df_vacancies['salary'].apply(
                lambda x: x.get('to', None) if isinstance(x, dict) else None
            )
            df_vacancies['salary.currency'] = df_vacancies['salary'].apply(
                lambda x: x.get('currency', 'RUR') if isinstance(x, dict) else 'RUR'
            )
        
        
    except Exception as e:
        print(f"Ошибка при загрузке датасета: {e}")
        # Если не удалось загрузить, используем демо-данные
        sample_data = {
            'name': ['Разработчик Python', 'Аналитик данных', 'Менеджер проектов'],
            'description': ['Разработка на Python и Django', 'Анализ данных с использованием Python и SQL', 'Управление IT проектами'],
            'salary.from': [100000, 80000, 120000],
            'salary.to': [150000, 120000, 180000],
            'experience.name': ['От 1 года до 3 лет', 'От 3 до 6 лет', 'Более 6 лет'],
            'area.name': ['Москва', 'Санкт-Петербург', 'Москва'],
            'schedule.name': ['Полный день', 'Полный день', 'Удаленная работа'],
            'alternate_url': ['http://example.com/1', 'http://example.com/2', 'http://example.com/3'],
            'employer.name': ['Компания А', 'Компания Б', 'Компания В'],
            'professional_roles': [{'name': 'Разработка'}, {'name': 'Аналитика'}, {'name': 'Менеджмент'}]
        }
        
        df_vacancies = pd.DataFrame(df_vacancies)
    
    root = tk.Tk()
    app = JobSearchSystem(root, df_vacancies)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, filedialog
from tkinter.scrolledtext import ScrolledText
import pandas as pd
import numpy as np
import re
import os
import json
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


class JobSearchSystem:
    def __init__(self, root, vacancies_df):
        self.root = root
        self.root.title("Система поиска вакансий")
        self.root.geometry("900x650")
        
        # Загрузка данных
        self.vacancies_df = vacancies_df
        self.preprocess_data()
        
        # Инициализация моделей
        self.init_models()
        
        # Переменные интерфейса
        self.job_title = tk.StringVar()
        self.resume_text = ""
        self.experience = tk.StringVar(value="Нет опыта")
        self.city = tk.StringVar()
        self.unemployed = tk.BooleanVar(value=False)
        self.same_field = tk.BooleanVar(value=True)
        self.desired_job = tk.StringVar()
        self.desired_salary = tk.StringVar()
        self.employment_type = tk.StringVar(value="Полный день")
        self.work_format = tk.StringVar(value="Не важно")
        
        # Создание интерфейса
        self.create_widgets()
    
    def preprocess_data(self):
        """Предварительная обработка данных"""
        # Заполнение пропущенных значений
        self.vacancies_df['salary.from'] = self.vacancies_df['salary.from'].fillna(0)
        self.vacancies_df['salary.to'] = self.vacancies_df['salary.to'].fillna(0)
        self.vacancies_df['experience.name'] = self.vacancies_df['experience.name'].fillna('Нет опыта')
        
        # Очистка текстовых полей
        text_columns = ['name', 'description']
        for col in text_columns:
            if col in self.vacancies_df.columns:
                self.vacancies_df[col] = self.vacancies_df[col].fillna('').apply(self.clean_text)
        
        # Создаем комбинированный текст для анализа
        self.vacancies_df['combined_text'] = (
            self.vacancies_df.get('name', '') + " " + 
            self.vacancies_df.get('description', '')
        ).apply(self.clean_text)
        
        # Нормализация зарплат
        self.vacancies_df['salary_mid'] = (self.vacancies_df['salary.from'] + self.vacancies_df['salary.to']) / 2
        
        # Подготовка данных для Item-Based CF
        self.prepare_item_based_data()
        
        # Подготовка данных для LSTM
        self.prepare_lstm_data()
    
    def prepare_lstm_data(self):
        """Подготовка данных для LSTM модели"""
        # Создаем категории на основе названий должностей
        self.vacancies_df['category'] = self.vacancies_df['name'].apply(
            lambda x: self.detect_category(x)
        )
        
        # Кодируем категории
        self.label_encoder = LabelEncoder()
        self.vacancies_df['category_encoded'] = self.label_encoder.fit_transform(
            self.vacancies_df['category']
        )
        
        # Сохраняем количество классов
        self.num_classes = len(self.label_encoder.classes_)
    
    def detect_category(self, job_title):
        """Определение категории на основе названия должности"""
        if not isinstance(job_title, str):
            return "Другое"
            
        job_title = job_title.lower()
        if any(word in job_title for word in ['разработчик', 'developer', 'программист', 'engineer']):
            return 'Разработка'
        elif any(word in job_title for word in ['аналитик', 'analyst', 'data scientist']):
            return 'Аналитика'
        elif any(word in job_title for word in ['менеджер', 'manager', 'руководитель']):
            return 'Менеджмент'
        elif any(word in job_title for word in ['дизайн', 'designer']):
            return 'Дизайн'
        else:
            return 'Другое'
    def prepare_item_based_data(self):
        """Подготовка данных для Item-Based Collaborative Filtering"""
        # Создаем матрицу признаков для вакансий
        self.job_features = pd.get_dummies(
            self.vacancies_df[['experience.name', 'schedule.name']], 
            columns=['experience.name', 'schedule.name'])

        # Добавляем нормализованную зарплату
        self.job_features['salary_norm'] = (
            self.vacancies_df['salary_mid'] - self.vacancies_df['salary_mid'].min()
        ) / (self.vacancies_df['salary_mid'].max() - self.vacancies_df['salary_mid'].min())

        # Заполняем пропуски
        self.job_features = self.job_features.fillna(0)

        # Создаем словарь для быстрого поиска индексов по названию должности
        # Используем очищенные названия должностей
        self.job_title_to_index = {
            self.clean_text(title): idx 
            for idx, title in enumerate(self.vacancies_df['name'])
        }

        # Создаем матрицу сходства для Item-Based CF
        self.item_similarity = 1 - pairwise_distances(
            self.job_features, 
            metric='cosine'
        )   

    
    def clean_text(self, text):
        """Очистка текста"""
        if not isinstance(text, str):
            return ""
        text = text.lower()
        text = re.sub(r'[^\w\s]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        return text
    
    def init_models(self):
        """Инициализация моделей машинного обучения"""
        self.lstm_model = None
        self.tokenizer = Tokenizer(num_words=10000)
        
        # Обучаем токенизатор на всех текстах
        texts = self.vacancies_df['combined_text'].tolist()
        self.tokenizer.fit_on_texts(texts) 
        
        # Пытаемся загрузить предобученную модель
        try:
            if os.path.exists('lstm_model.h5'):
                self.lstm_model = load_model('lstm_model.h5')
                # Загружаем историю обучения для отображения графиков
                if os.path.exists('training_history.json'):
                    with open('training_history.json', 'r') as f:
                        self.training_history = json.load(f)
            else:
                self.train_lstm_model()
        except Exception as e:
            print(f"Ошибка при загрузке модели: {e}")
            self.train_lstm_model()
        
        # Подготовка модели рекомендаций
        self.prepare_recommendation_model()
    
    def train_lstm_model(self):
        """Обучение LSTM модели на текстах вакансий"""
        texts = self.vacancies_df['combined_text'].tolist()
        labels = self.vacancies_df['category_encoded'].values
        
        # Преобразование текста в последовательности
        sequences = self.tokenizer.texts_to_sequences(texts)
        padded_sequences = pad_sequences(sequences, maxlen=200)
        
        # Разделение на обучающую и тестовую выборки
        X_train, X_test, y_train, y_test = train_test_split(
            padded_sequences, labels, test_size=0.2, random_state=42
        )
        
        # Преобразование меток в one-hot encoding
        y_train = to_categorical(y_train, num_classes=self.num_classes)
        y_test = to_categorical(y_test, num_classes=self.num_classes)
        
        # Создание модели
        self.lstm_model = Sequential([
            Embedding(input_dim=10000, output_dim=128, input_length=200),
            LSTM(128, dropout=0.2, recurrent_dropout=0.2, return_sequences=True),
            LSTM(64, dropout=0.2, recurrent_dropout=0.2),
            Dense(64, activation='relu'),
            Dropout(0.5),
            Dense(self.num_classes, activation='softmax')
        ])
        
        self.lstm_model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        # Коллбэки
        callbacks = [
            EarlyStopping(patience=3, restore_best_weights=True),
            ModelCheckpoint('lstm_model.h5', save_best_only=True)
        ]
        
        # Обучение модели
        history = self.lstm_model.fit(
            X_train,
            y_train,
            batch_size=64,
            epochs=20,
            validation_data=(X_test, y_test),
            callbacks=callbacks
        )
        
        # Сохраняем историю обучения
        self.training_history = history.history
        with open('training_history.json', 'w') as f:
            json.dump(self.training_history, f)
        
        # Визуализация обучения
        self.plot_training_history()
    
    def plot_training_history(self):
        """Визуализация процесса обучения модели"""
        plt.figure(figsize=(12, 5))
        
        # График точности
        plt.subplot(1, 2, 1)
        plt.plot(self.training_history['accuracy'], label='Точность на обучении')
        plt.plot(self.training_history['val_accuracy'], label='Точность на валидации')
        plt.title('Точность модели')
        plt.ylabel('Точность')
        plt.xlabel('Эпоха')
        plt.legend()
        
        # # График потерь
        plt.subplot(1, 2, 2)
        plt.plot(self.training_history['loss'], label='Потери на обучении')
        plt.plot(self.training_history['val_loss'], label='Потери на валидации')
        plt.title('Потери модели')
        plt.ylabel('Потери')
        plt.xlabel('Эпоха')
        plt.legend()
        
        plt.tight_layout()
        plt.savefig('training_history.png')
        plt.close()
    
    def prepare_recommendation_model(self):
        """Подготовка модели рекомендаций на основе TF-IDF и косинусного сходства"""
        # Создаем TF-IDF матрицу для текстов вакансий
        self.tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
        self.tfidf_matrix = self.tfidf_vectorizer.fit_transform(self.vacancies_df['combined_text'])
        
        # Создаем матрицу сходства для Item-Based CF
        self.item_similarity = 1 - pairwise_distances(
            self.job_features, 
            metric='cosine'
        )
        
        # Модель KNN для рекомендаций
        self.knn_model = NearestNeighbors(n_neighbors=10, metric='cosine')
        self.knn_model.fit(self.tfidf_matrix)
    
    def create_widgets(self):
        """Создание интерфейса пользователя"""
        main_frame = ttk.Frame(self.root, padding="10")
        main_frame.pack(fill=BOTH, expand=True)

        # Блок названия должности
        ttk.Label(main_frame, text="Название текущей/желаемой должности:").grid(row=0, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.job_title, width=50).grid(row=0, column=1, sticky=tk.W, pady=5)

        # Блок резюме
        resume_frame = ttk.LabelFrame(main_frame, text="Резюме", padding=10)
        resume_frame.grid(row=1, column=0, columnspan=2, sticky=tk.EW, pady=10)

        ttk.Button(resume_frame, text="Ввести резюме", command=self.open_resume_editor).pack(side=tk.LEFT, padx=5)
        ttk.Button(resume_frame, text="Загрузить из файла", command=self.load_resume_file).pack(side=tk.LEFT, padx=5)

        # Блок опыта работы
        ttk.Label(main_frame, text="Опыт работы:").grid(row=2, column=0, sticky=tk.W, pady=5)
        experience_options = ["Нет опыта", "От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"]
        ttk.Combobox(main_frame, textvariable=self.experience, values=experience_options, state="readonly").grid(
            row=2, column=1, sticky=tk.W, pady=5)

        # Блок города
        ttk.Label(main_frame, text="Город:").grid(row=3, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.city, width=30).grid(row=3, column=1, sticky=tk.W, pady=5)

        # Чекбокс поиска в той же сфере
        ttk.Checkbutton(main_frame, text="Искать в той же сфере деятельности", variable=self.same_field,
                       command=self.toggle_job_field).grid(row=4, column=0, columnspan=2, sticky=tk.W, pady=5)

        # Блок типа занятости
        ttk.Label(main_frame, text="Тип занятости:").grid(row=5, column=0, sticky=tk.W, pady=5)
        employment_options = ["Полный день", "Удаленная работа", "Не важно"]
        ttk.Combobox(main_frame, textvariable=self.employment_type, values=employment_options, state="readonly", width=20).grid(row=5, column=1, sticky=tk.W, pady=5)

        # Блок формата работы
        ttk.Label(main_frame, text="Формат работы:").grid(row=6, column=0, sticky=tk.W, pady=5)
        format_options = ["На месте работодателя", "Удалённо", "Гибрид", "Не важно"]
        ttk.Combobox(main_frame, textvariable=self.work_format, values=format_options, state="readonly", width=20).grid(row=6, column=1, sticky=tk.W, pady=5)

        # Блок желаемой зарплаты
        ttk.Label(main_frame, text="Желаемая зарплата (руб.):").grid(row=7, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.desired_salary, width=15).grid(row=7, column=1, sticky=tk.W, pady=5)

        # Чекбокс безработного статуса
        ttk.Checkbutton(main_frame, text="Я в настоящее время безработный", variable=self.unemployed).grid(
            row=8, column=0, columnspan=2, sticky=tk.W, pady=5)

        # Кнопка поиска
        ttk.Button(main_frame, text="Найти вакансии", command=self.search_jobs).grid(
            row=9, column=0, columnspan=2, pady=20)

        # Кнопка выхода
        ttk.Button(main_frame, text="Выход", command=self.root.quit).grid(
            row=10, column=0, columnspan=2, pady=10)
    
    def toggle_job_field(self):
        """Обработчик изменения чекбокса 'Искать в той же сфере'"""
        if not self.same_field.get():
            self.open_desired_job_dialog()
    
    def open_desired_job_dialog(self):
        """Открытие диалога для указания желаемой должности"""
        dialog = tk.Toplevel(self.root)
        dialog.title("Укажите желаемую должность")
        dialog.geometry("400x200")

        ttk.Label(dialog, text="Введите желаемую должность:").pack(pady=10)
        entry = ttk.Entry(dialog, textvariable=self.desired_job, width=40)
        entry.pack(pady=5)
        entry.focus_set()

        def save_and_close():
            if not self.desired_job.get():
                messagebox.showerror("Ошибка", "Пожалуйста, введите желаемую должность")
                return
            dialog.destroy()

        ttk.Button(dialog, text="Сохранить", command=save_and_close).pack(pady=15)
    
    def open_resume_editor(self):
        """Открытие редактора резюме"""
        editor = tk.Toplevel(self.root)
        editor.title("Редактор резюме")
        editor.geometry("600x400")

        text_area = ScrolledText(editor, wrap=tk.WORD, width=70, height=20)
        text_area.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)

        if self.resume_text:
            text_area.insert(tk.END, self.resume_text)

        def save_resume():
            self.resume_text = text_area.get("1.0", tk.END)
            editor.destroy()
            messagebox.showinfo("Сохранено", "Резюме успешно сохранено")

        ttk.Button(editor, text="Сохранить", command=save_resume).pack(pady=5)
    
    def load_resume_file(self):
        """Загрузка резюме из файла"""
        file_path = filedialog.askopenfilename(
            filetypes=[("Текстовые файлы", "*.txt"), ("Документы Word", "*.docx"), ("PDF", "*.pdf")])
        if file_path:
            try:
                if file_path.endswith('.txt'):
                    with open(file_path, 'r', encoding='utf-8') as file:
                        self.resume_text = file.read()
                else:
                    self.resume_text = f"Файл {file_path} загружен (необходима обработка для извлечения текста)"
                messagebox.showinfo("Успех", "Резюме успешно загружено")
            except Exception as e:
                messagebox.showerror("Ошибка", f"Не удалось загрузить файл:\n{str(e)}")
# ... (предыдущий код остается без изменений до метода search_jobs)

    def search_jobs(self):
        """Поиск вакансий"""
        # Проверка обязательных полей
        if not self.job_title.get() and not self.same_field.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите название должности или выберите 'Искать в той же сфере деятельности'")
            return

        if not self.resume_text:
            messagebox.showerror("Ошибка", "Пожалуйста, введите или загрузите резюме")
            return

        if not self.city.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите город")
            return

        if not self.desired_salary.get() or not self.desired_salary.get().isdigit():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите валидную желаемую зарплату")
            return

        # Проверка для случая, когда снята галочка "Искать в той же сфере"
        if not self.same_field.get() and not self.desired_job.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите желаемую должность")
            return

        # Подготовка данных для поиска
        search_data = {
            "job_title": self.job_title.get(),
            "resume_text": self.resume_text,
            "experience": self.experience.get(),
            "city": self.city.get(),
            "same_field": self.same_field.get(),
            "desired_job": self.desired_job.get() if not self.same_field.get() else "",
            "employment_type": self.employment_type.get(),
            "work_format": self.work_format.get(),
            "desired_salary": int(self.desired_salary.get()),
            "unemployed": self.unemployed.get()
        }

        # Поиск вакансий
        if self.same_field.get():
            # Используем текст резюме для поиска в той же сфере
            filtered_jobs = self.find_relevant_jobs_by_resume(search_data)
        else:
            # Используем Item-Based CF для поиска в другой сфере
            filtered_jobs = self.find_jobs_with_item_based_cf(search_data)
        
        # Отображение результатов
        self.display_results(filtered_jobs)
        
        
    def find_jobs_with_item_based_cf(self, search_data):
        """Поиск вакансий с использованием Item-Based Collaborative Filtering и текстового сходства"""
        # Определяем целевую должность для поиска
        target_job = search_data['desired_job'] if search_data['desired_job'] else search_data['job_title']
        target_job_clean = self.clean_text(target_job)

        if not target_job_clean:
            return pd.DataFrame()

        # 1. Поиск по точному или частичному совпадению в названии
        similar_indices = []
        for idx, title in enumerate(self.vacancies_df['name_desc']):
            clean_title = self.clean_text(str(title))
            if target_job_clean in clean_title or clean_title in target_job_clean:
                similar_indices.append(idx)
                if len(similar_indices) >= 5:
                    break

        # 2. Если не нашли по названию, ищем по описанию
        if not similar_indices:
            for idx, desc in enumerate(self.vacancies_df['description']):
                clean_desc = self.clean_text(str(desc))
                if target_job_clean in clean_desc:
                                   similar_indices.append(idx)
                if len(similar_indices) >= 3:
                        break

        # 3. Если совсем не нашли, используем TF-IDF
        if not similar_indices:
            target_vector = self.tfidf_vectorizer.transform([target_job_clean])
            similarities = cosine_similarity(target_vector, self.tfidf_matrix)[0]
            similar_indices = np.argsort(-similarities)[:10].tolist()

        if not similar_indices:
            return pd.DataFrame()

        # Получаем рекомендации на основе сходства
        all_recommendations = []
        for job_idx in similar_indices:
            if job_idx < len(self.item_similarity):
                recs = np.argsort(-self.item_similarity[job_idx])[1:6]
                all_recommendations.extend(recs)

        # Удаляем дубликаты и проверяем индексы
        unique_recommendations = list(set(
            idx for idx in all_recommendations 
            if idx < len(self.vacancies_df)
        ))

        if not unique_recommendations:
            return pd.DataFrame()

        filtered_jobs = self.vacancies_df.iloc[unique_recommendations].copy()



        # Фильтр по опыту работы
        experience_map = {
            "Нет опыта": ["Нет опыта"],
            "От 1 года до 3 лет": ["От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"],
            "От 3 до 6 лет": ["От 3 до 6 лет", "Более 6 лет"],
            "Более 6 лет": ["Более 6 лет"]
        }
        if search_data['experience'] in experience_map:
            filtered_jobs = filtered_jobs[
                filtered_jobs['experience.name'].isin(experience_map[search_data['experience']])
            ]

        # Фильтр по зарплате
        desired_salary = search_data['desired_salary']
        if 'salary.from' in filtered_jobs.columns and 'salary.to' in filtered_jobs.columns:
            filtered_jobs = filtered_jobs[
                (filtered_jobs['salary.from'].fillna(0) <= desired_salary * 1.5) | 
                (filtered_jobs['salary.to'].fillna(float('inf')) >= desired_salary * 0.7)
            ]

        # Вычисляем сходство
        if not filtered_jobs.empty:
            job_texts = filtered_jobs['name'] + " " + filtered_jobs['description']
            job_vectors = self.tfidf_vectorizer.transform(job_texts)
            target_vector = self.tfidf_vectorizer.transform([target_job_clean])
            similarities = cosine_similarity(target_vector, job_vectors)[0]
            filtered_jobs['similarity'] = similarities
            filtered_jobs = filtered_jobs.sort_values('similarity', ascending=False)

        return filtered_jobs.head(10)



    def find_relevant_jobs_by_resume(self, search_data):
        """Поиск релевантных вакансий по тексту резюме с использованием TF-IDF и LSTM"""
        # Базовые фильтры
        filtered_df = self.vacancies_df.copy()
        
        # Фильтр по городу
        if search_data['city']:
            filtered_df = filtered_df[filtered_df['area.name_desc'].str.contains(
                search_data['city'], case=False, na=False)]
        
        # Фильтр по опыту работы
        experience_map = {
            "Нет опыта": ["Нет опыта"],
            "От 1 года до 3 лет": ["От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"],
            "От 3 до 6 лет": ["От 3 до 6 лет", "Более 6 лет"],
            "Более 6 лет": ["Более 6 лет"]}
        if search_data['experience'] in experience_map:
            filtered_df = filtered_df[filtered_df['experience.name'].isin(experience_map[search_data['experience']])]
        
        # Фильтр по типу занятости
        if search_data['employment_type'] != "Не важно":
            filtered_df = filtered_df[filtered_df['schedule.name'].str.contains(
                search_data['employment_type'], case=False, na=False)]
        
        # Фильтр по желаемой зарплате
        desired_salary = search_data['desired_salary']
        filtered_df = filtered_df[
            (filtered_df['salary.from'] <= desired_salary) | 
            (filtered_df['salary.to'] >= desired_salary)]
        
        # Если после базовых фильтров осталось слишком много вакансий, применяем TF-IDF и LSTM
        if len(filtered_df) > 0:
            # Получаем вектор резюме для TF-IDF
            resume_vector = self.tfidf_vectorizer.transform([search_data['resume_text']])
            
            # Получаем векторы вакансий
            job_vectors = self.tfidf_vectorizer.transform(filtered_df['combined_text'])
            
            # Вычисляем косинусное сходство для TF-IDF
            similarities_tfidf = cosine_similarity(resume_vector, job_vectors)[0]
            
            # Получаем предсказания категории LSTM для резюме
            resume_sequence = self.tokenizer.texts_to_sequences([search_data['resume_text']])
            resume_padded = pad_sequences(resume_sequence, maxlen=200)
            resume_category_prob = self.lstm_model.predict(resume_padded)[0]
            
            # Получаем предсказания категорий для вакансий
            job_sequences = self.tokenizer.texts_to_sequences(filtered_df['combined_text'].tolist())
            job_padded = pad_sequences(job_sequences, maxlen=200)
            job_categories_prob = self.lstm_model.predict(job_padded)
            
            # Вычисляем сходство категорий (косинусное сходство между векторами вероятностей)
            similarities_lstm = cosine_similarity(
                resume_category_prob.reshape(1, -1), 
                job_categories_prob)[0]
            
            # Комбинируем оценки (можно настроить веса)
            combined_similarity = 0.6 * similarities_tfidf + 0.4 * similarities_lstm
            filtered_df['similarity'] = combined_similarity
            filtered_df = filtered_df.sort_values('similarity', ascending=False)
            
            # Ограничиваем количество результатов
            filtered_df = filtered_df.head(10)
        
        return filtered_df
    
    
    def display_results(self, filtered_jobs):
        """Отображение результатов поиска"""
        if filtered_jobs.empty:
            messagebox.showinfo("Результаты поиска", "К сожалению, не найдено подходящих вакансий.")
        else:
            # Создаем окно с результатами
            results_window = tk.Toplevel(self.root)
            results_window.title("Результаты поиска вакансий")
            results_window.geometry("900x600")
            
            # Создаем Treeview для отображения таблицы
            tree = ttk.Treeview(results_window, columns=('name', 'company', 'salary', 'experience', 'similarity'), show='headings')
            tree.heading('name', text='Должность')
            tree.heading('company', text='Компания')
            tree.heading('salary', text='Зарплата')
            tree.heading('experience', text='Опыт')
            tree.heading('similarity', text='Сходство')
            
            tree.column('name', width=200)
            tree.column('company', width=150)
            tree.column('salary', width=100)
            tree.column('experience', width=100)
            tree.column('similarity', width=80)
            
            # Добавляем данные
            for _, row in filtered_jobs.iterrows():
                salary_from = row['salary.from']
                salary_to = row['salary.to']
                
                if pd.notna(salary_from) and pd.notna(salary_to):
                    salary_str = f"{int(salary_from)}-{int(salary_to)}"
                elif pd.notna(salary_from):
                    salary_str = f"от {int(salary_from)}"
                elif pd.notna(salary_to):
                    salary_str = f"до {int(salary_to)}"
                else:
                    salary_str = "Не указано"
                
                similarity = row.get('similarity', 0)
                similarity_str = f"{similarity:.2f}" if isinstance(similarity, (float, np.float64)) else "N/A"
                
                tree.insert('', tk.END, values=(
                    row.get('name', 'Не указано'),
                    row.get('employer.name', 'Не указано'),
                    salary_str,
                    row.get('experience.name', 'Не указано'),
                    similarity_str))
            
            # Добавляем полосу прокрутки
            scrollbar = ttk.Scrollbar(results_window, orient="vertical", command=tree.yview)
            tree.configure(yscrollcommand=scrollbar.set)
            scrollbar.pack(side="right", fill="y")
            tree.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)
            
            # Кнопка закрытия
            ttk.Button(results_window, text="Закрыть", command=results_window.destroy).pack(pady=10)


if __name__ == "__main__":
    # Загрузка реального датасета
    try:
        # Пытаемся загрузить данные из CSV файла
        df_vacancies = pd.read_csv('vacancies.csv', low_memory=False)
        
        # Если в данных есть JSON-столбцы, преобразуем их
        json_columns = ['employment', 'experience', 'salary']
        for col in json_columns:
            if col in df_vacancies.columns:
                df_vacancies[col] = df_vacancies[col].apply(
                    lambda x: json.loads(x.replace("'", "\"")) if isinstance(x, str) else x)
        
        # Извлекаем вложенные данные
        if 'experience' in df_vacancies.columns:
            df_vacancies['experience.name'] = df_vacancies['experience'].apply(
                lambda x: x.get('name', 'Нет опыта') if isinstance(x, dict) else 'Нет опыта')
        
        if 'employment' in df_vacancies.columns:
            df_vacancies['schedule.name'] = df_vacancies['employment'].apply(
                lambda x: x.get('name', '') if isinstance(x, dict) else ''
            )
        
        if 'salary' in df_vacancies.columns:
            df_vacancies['salary.from'] = df_vacancies['salary'].apply(
                lambda x: x.get('from', None) if isinstance(x, dict) else None
            )
            df_vacancies['salary.to'] = df_vacancies['salary'].apply(
                lambda x: x.get('to', None) if isinstance(x, dict) else None
            )
            df_vacancies['salary.currency'] = df_vacancies['salary'].apply(
                lambda x: x.get('currency', 'RUR') if isinstance(x, dict) else 'RUR'
            )
        
    except Exception as e:
        print(f"Ошибка при загрузке датасета: {e}")
        # Если не удалось загрузить, используем демо-данные
        sample_data_data = {
            'name': ['Разработчик Python', 'Аналитик данных', 'Менеджер проектов'],
            'description': ['Разработка на Python и Django', 'Анализ данных с использованием Python и SQL', 'Управление IT проектами'],
            'salary.from': [100000, 80000, 120000],
            'salary.to': [150000, 120000, 180000],
            'experience.name': ['От 1 года до 3 лет', 'От 3 до 6 лет', 'Более 6 лет'],
            'area.name_desc': ['Москва', 'Санкт-Петербург', 'Москва'],
            'schedule.name': ['Полный день', 'Полный день', 'Удаленная работа'],
            'alternate_url': ['http://example.com/1', 'http://example.com/2', 'http://example.com/3'],
            'employer.name': ['Компания А', 'Компания Б', 'Компания В']
        }
        
        df_vacancies = pd.DataFrame(sample_data)
    
    root = tk.Tk()
    app = JobSearchSystem(root, df_vacancies)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, filedialog
from tkinter.scrolledtext import ScrolledText
import pandas as pd
import numpy as np
import re
import os
import json
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from tkinter.constants import BOTH


class JobSearchSystem:
    def __init__(self, root, vacancies_df):
        self.root = root
        self.root.title("Система поиска вакансий")
        self.root.geometry("900x650")
        
        # Загрузка данных
        self.vacancies_df = vacancies_df
        self.preprocess_data()
        
        # Инициализация моделей
        self.init_models()
        
        # Переменные интерфейса
        self.job_title = tk.StringVar()
        self.resume_text = ""
        self.experience = tk.StringVar(value="Нет опыта")
        self.city = tk.StringVar()
        self.unemployed = tk.BooleanVar(value=False)
        self.same_field = tk.BooleanVar(value=True)
        self.desired_job = tk.StringVar()
        self.desired_salary = tk.StringVar()
        self.employment_type = tk.StringVar(value="Полный день")
        self.work_format = tk.StringVar(value="Не важно")
        
        # Создание интер интерфейса
        self.create_widgets()
    
    def preprocess_data(self):
        """Предварительная обработка данных"""
        # Заполнение пропущенных значений
        self.vacancies_df['salary.from'] = self.vacancies_df['salary.from'].fillna(0)
        self.vacancies_df['salary.to'] = self.vacancies_df['salary.to'].fillna(0)
        self.vacancies_df['experience.name'] = self.vacancies_df['experience.name'].fillna('Нет опыта')
        
        # Очистка текстовых полей
        text_columns = ['name', 'processed_description_desc']
        for col in text_columns:
            if col in self.vacancies_df.columns:
                self.vacancies_df[col] = self.vacancies_df[col].fillna('').apply(self.clean_text)
        
        # Создаем комбинированный текст для анализа
        self.vacancies_df['combined_text'] = (
            self.vacancies_df.get('name', '') + " " + 
            self.vacancies_df.get('processed_description_desc', '')
        ).apply(self.clean_text)
        
        # Нормализация зарплат
        self.vacancies_df['salary_mid'] = (self.vacancies_df['salary.from'] + self.vacancies_df['salary.to']) / 2
        
        # Подготовка данных для Item-Based CF
        self.prepare_item_based_data()
        
        # Подготовка данных для LSTM
        self.prepare_lstm_data()
    
    def prepare_lstm_data(self):
        """Подготовка данных для LSTM модели"""
        # Создаем категории на основе названий должностей
        self.vacancies_df['category'] = self.vacancies_df['name'].apply(
            lambda x: self.detect_category(x)
        )
        
        # Кодируем категории
        self.label_encoder = LabelEncoder()
        self.vacancies_df['category_encoded'] = self.label_encoder.fit_transform(
            self.vacancies_df['category']
        )
        
        # Сохраняем количество классов
        self.num_classes = len(self.label_encoder.classes_)
    
    def detect_category(self, job_title):
        """Определение категории на основе названия должности"""
        if not isinstance(job_title, str):
            return "Другое"
            
        job_title = job_title.lower()
        if any(word in job_title for word in ['разработчик', 'developer', 'программист', 'engineer']):
            return 'Разработка'
        elif any(word in job_title for word in ['аналитик', 'analyst', 'data scientist']):
            return 'Аналитика'
        elif any(word in job_title for word in ['менеджер', 'manager', 'руководитель']):
            return 'Менеджмент'
        elif any(word in job_title for word in ['дизайн', 'designer']):
            return 'Дизайн'
        else:
            return 'Другое'
    
    def prepare_item_based_data(self):
        """Подготовка данных для Item-Based Collaborative Filtering"""
        # Создаем матрицу признаков для вакансий
        self.job_features = pd.get_dummies(
            self.vacancies_df[['experience.name', 'schedule.name']], 
            columns=['experience.name', 'schedule.name'])

        # Добавляем нормализованную зарплату
        self.job_features['salary_norm'] = (
            self.vacancies_df['salary_mid'] - self.vacancies_df['salary_mid'].min()
        ) / (self.vacancies_df['salary_mid'].max() - self.vacancies_df['salary_mid'].min())

        # Заполняем пропуски
        self.job_features = self.job_features.fillna(0)

        # Создаем словарь для быстрого поиска индексов по названию должности
        self.job_title_to_index = {
            self.clean_text(title): idx 
            for idx, title in enumerate(self.vacancies_df['name'])
        }

        # Создаем матрицу сходства для Item-Based CF
        self.item_similarity = 1 - pairwise_distances(
            self.job_features, 
            metric='cosine'
        )
    
    def clean_text(self, text):
        """Очистка текста"""
        if not isinstance(text, str):
            return ""
        text = text.lower()
        text = re.sub(r'[^\w\s]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        return text
    
    def init_models(self):
        """Инициализация моделей машинного обучения"""
        self.lstm_model = None
        self.tokenizer = Tokenizer(num_words=10000)
        
        # Обучаем токенизатор на всех текстах
        texts = self.vacancies_df['combined_text'].tolist()
        self.tokenizer.fit_on_texts(texts) 
        
        # Пытаемся загрузить предобученную модель
        try:
            if os.path.exists('lstm_model.h5'):
                self.lstm_model = load_model('lstm_model.h5')
                # Загружаем историю обучения для отображения графиков
                if os.path.exists('training_history.json'):
                    with open('training_history.json', 'r') as f:
                        self.training_history = json.load(f)
            else:
                self.train_lstm_model()
        except Exception as e:
            print(f"Ошибка при загрузке модели: {e}")
            self.train_lstm_model()
        
        # Подготовка модели рекомендаций
        self.prepare_recommendation_model()
    
    def train_lstm_model(self):
        """Обучение LSTM модели на текстах вакансий"""
        texts = self.vacancies_df['combined_text'].tolist()
        labels = self.vacancies_df['category_encoded'].values
        
        # Преобразование текста в последовательности
        sequences = self.tokenizer.texts_to_sequences(texts)
        padded_sequences = pad_sequences(sequences, maxlen=200)
        
        # Разделение на обучающую и тестовую выборки
        X_train, X_test, y_train, y_test = train_test_split(
            padded_sequences, labels, test_size=0.2, random_state=42
        )
        
        # Преобразование меток в one-hot encoding
        y_train = to_categorical(y_train, num_classes=self.num_classes)
        y_test = to_categorical(y_test, num_classes=self.num_classes)
        
        # Создание модели
        self.lstm_model = Sequential([
            Embedding(input_dim=10000, output_dim=128, input_length=200),
            LSTM(128, dropout=0.2, recurrent_dropout=0.2, return_sequences=True),
            LSTM(64, dropout=0.2, recurrent_dropout=0.2),
            Dense(64, activation='relu'),
            Dropout(0.5),
            Dense(self.num_classes, activation='softmax')
        ])
        
        self.lstm_model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        # Коллбэки
        callbacks = [
            EarlyStopping(patience=3, restore_best_weights=True),
            ModelCheckpoint('lstm_model.h5', save_best_only=True)
        ]
        
        # Обучение модели
        history = self.lstm_model.fit(
            X_train,
            y_train,
            batch_size=64,
            epochs=20,
            validation_data=(X_test, y_test),
            callbacks=callbacks
        )
        
        # Сохраняем историю обучения
        self.training_history = history.history
        with open('training_history.json', 'w') as f:
            json.dump(self.training_history, f)
        
        # Визуализация обучения
        self.plot_training_history()
    
    def plot_training_history(self):
        """Визуализация процесса обучения модели"""
        plt.figure(figsize=(12, 5))
        
        # График точности
        plt.subplot(1, 2, 1)
        plt.plot(self.training_history['accuracy'], label='Точность на обучении')
        plt.plot(self.training_history['val_accuracy'], label='Точность на валидации')
        plt.title('Точность модели')
        plt.ylabel('Точность')
        plt.xlabel('Эпоха')
        plt.legend()
        
        # График потерь
        plt.subplot(1, 2, 2)
        plt.plot(self.training_history['loss'], label='Потери на обучении')
        plt.plot(self.training_history['val_loss'], label='Потери на валидации')
        plt.title('Потери модели')
        plt.ylabel('Потери')
        plt.xlabel('Эпоха')
        plt.legend()
        
        plt.tight_layout()
        plt.savefig('training_history.png')
        plt.close()
    
    def prepare_recommendation_model(self):
        """Подготовка модели рекомендаций на основе TF-IDF и косинусного сходства"""
        # Создаем TF-IDF матрицу для текстов вакансий
        self.tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
        self.tfidf_matrix = self.tfidf_vectorizer.fit_transform(self.vacancies_df['combined_text'])
        
        # Модель KNN для рекомендаций
        self.knn_model = NearestNeighbors(n_neighbors=10, metric='cosine')
        self.knn_model.fit(self.tfidf_matrix)
    
    def create_widgets(self):
        """Создание интерфейса пользователя"""
        main_frame = ttk.Frame(self.root, padding="10")
        main_frame.pack(fill=BOTH, expand=True)

        # Блок названия должности
        ttk.Label(main_frame, text="Название текущей/желаемой должности:").grid(row=0, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.job_title, width=50).grid(row=0, column=1, sticky=tk.W, pady=5)

        # Блок резюме
        resume_frame = ttk.LabelFrame(main_frame, text="Резюме", padding=10)
        resume_frame.grid(row=1, column=0, columnspan=2, sticky=tk.EW, pady=10)

        ttk.Button(resume_frame, text="Ввести резюме", command=self.open_resume_editor).pack(side=tk.LEFT, padx=5)
        ttk.Button(resume_frame, text="Загрузить из файла", command=self.load_resume_file).pack(side=tk.LEFT, padx=5)

        # Блок опыта работы
        ttk.Label(main_frame, text="Опыт работы:").grid(row=2, column=0, sticky=tk.W, pady=5)
        experience_options = ["Нет опыта", "От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"]
        ttk.Combobox(main_frame, textvariable=self.experience, values=experience_options, state="readonly").grid(
            row=2, column=1, sticky=tk.W, pady=5)

        # Блок города
        ttk.Label(main_frame, text="Город:").grid(row=3, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.city, width=30).grid(row=3, column=1, sticky=tk.W, pady=5)

        # Чекбокс поиска в той же сфере
        ttk.Checkbutton(main_frame, text="Искать в той же сфере деятельности", variable=self.same_field,
                       command=self.toggle_job_field).grid(row=4, column=0, columnspan=2, sticky=tk.W, pady=5)

        # Блок типа занятости
        ttk.Label(main_frame, text="Тип занятости:").grid(row=5, column=0, sticky=tk.W, pady=5)
        employment_options = ["Полный день", "Удаленная работа", "Не важно"]
        ttk.Combobox(main_frame, textvariable=self.employment_type, values=employment_options, state="readonly", width=20).grid(row=5, column=1, sticky=tk.W, pady=5)

        # Блок формата работы
        ttk.Label(main_frame, text="Формат работы:").grid(row=6, column=0, sticky=tk.W, pady=5)
        format_options = ["На месте работодателя", "Удалённо", "Гибрид", "Не важно"]
        ttk.Combobox(main_frame, textvariable=self.work_format, values=format_options, state="readonly", width=20).grid(row=6, column=1, sticky=tk.W, pady=5)

        # Блок желаемой зарплаты
        ttk.Label(main_frame, text="Желаемая зарплата (руб.):").grid(row=7, column=0, sticky=tk.W, pady=5)
        ttk.Entry(main_frame, textvariable=self.desired_salary, width=15).grid(row=7, column=1, sticky=tk.W, pady=5)

        # Чекбокс безработного статуса
        ttk.Checkbutton(main_frame, text="Я в настоящее время безработный", variable=self.unemployed).grid(
            row=8, column=0, columnspan=2, sticky=tk.W, pady=5)

        # Кнопка поиска
        ttk.Button(main_frame, text="Найти вакансии", command=self.search_jobs).grid(
            row=9, column=0, columnspan=2, pady=20)

        # Кнопка выхода
        ttk.Button(main_frame, text="Выход", command=self.root.quit).grid(
            row=10, column=0, columnspan=2, pady=10)
    
    def toggle_job_field(self):
        """Обработчик изменения чекбокса 'Искать в той же сфере'"""
        if not self.same_field.get():
            self.open_desired_job_dialog()
    
    def open_desired_job_dialog(self):
        """Открытие диалога для указания желаемой должности"""
        dialog = tk.Toplevel(self.root)
        dialog.title("Укажите желаемую должность")
        dialog.geometry("400x200")

        ttk.Label(dialog, text="Введите желаемую должность:").pack(pady=10)
        entry = ttk.Entry(dialog, textvariable=self.desired_job, width=40)
        entry.pack(pady=5)
        entry.focus_set()

        def save_and_close():
            if not self.desired_job.get():
                messagebox.showerror("Ошибка", "Пожалуйста, введите желаемую должность")
                return
            dialog.destroy()

        ttk.Button(dialog, text="Сохранить", command=save_and_close).pack(pady=15)
    
    def open_resume_editor(self):
        """Открытие редактора резюме"""
        editor = tk.Toplevel(self.root)
        editor.title("Редактор резюме")
        editor.geometry("600x400")

        text_area = ScrolledText(editor, wrap=tk.WORD, width=70, height=20)
        text_area.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)

        if self.resume_text:
            text_area.insert(tk.END, self.resume_text)

        def save_resume():
            self.resume_text = text_area.get("1.0", tk.END)
            editor.destroy()
            messagebox.showinfo("Сохранено", "Резюме успешно сохранено")

        ttk.Button(editor, text="Сохранить", command=save_resume).pack(pady=5)
    
    def load_resume_file(self):
        """Загрузка резюме из файла"""
        file_path = filedialog.askopenfilename(
            filetypes=[("Текстовые файлы", "*.txt"), ("Документы Word", "*.docx"), ("PDF", "*.pdf")])
        if file_path:
            try:
                if file_path.endswith('.txt'):
                    with open(file_path, 'r', encoding='utf-8') as file:
                        self.resume_text = file.read()
                else:
                    self.resume_text = f"Файл {file_path} загружен (необходима обработка для извлечения текста)"
                messagebox.showinfo("Успех", "Резюме успешно загружено")
            except Exception as e:
                messagebox.showerror("Ошибка", f"Не удалось загрузить файл:\n{str(e)}")
    
    def search_jobs(self):
        """Поиск вакансий"""
        # Проверка обязательных полей
        if not self.job_title.get() and not self.same_field.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите название должности или выберите 'Искать в той же сфере деятельности'")
            return

        if not self.resume_text:
            messagebox.showerror("Ошибка", "Пожалуйста, введите или загрузите резюме")
            return

        if not self.city.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите город")
            return

        if not self.desired_salary.get() or not self.desired_salary.get().isdigit():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите валидную желаемую зарплату")
            return

        # Проверка для случая, когда снята галочка "Искать в той же сфере"
        if not self.same_field.get() and not self.desired_job.get():
            messagebox.showerror("Ошибка", "Пожалуйста, укажите желаемую должность")
            return

        # Подготовка данных для поиска
        search_data = {
            "job_title": self.job_title.get(),
            "resume_text": self.resume_text,
            "experience": self.experience.get(),
            "city": self.city.get(),
            "same_field": self.same_field.get(),
            "desired_job": self.desired_job.get() if not self.same_field.get() else "",
            "employment_type": self.employment_type.get(),
            "work_format": self.work_format.get(),
            "desired_salary": int(self.desired_salary.get()),
            "unemployed": self.unemployed.get()
        }

        # Поиск вакансий
        if self.same_field.get():
            # Используем текст резюме для поиска в той же сфере
            filtered_jobs = self.find_relevant_jobs_by_resume(search_data)
        else:
            # Используем желаемую должность для поиска в другой сфере
            filtered_jobs = self.find_jobs_in_different_field(search_data)
        
        # Отображение результатов
        self.display_results(filtered_jobs)
    
    def find_jobs_in_different_field(self, search_data):
        """Поиск вакансий в другой сфере деятельности"""
        # Получаем желаемую должность
        target_job = search_data['desired_job']
        if not target_job:
            return pd.DataFrame()
        
        # Очищаем текст
        target_job_clean = self.clean_text(target_job)
        
        # Базовые фильтры
        filtered_df = self.vacancies_df.copy()
        
        # Фильтр по городу
        if search_data['city']:
            filtered_df = filtered_df[filtered_df['area.name_desc'].str.contains(
                search_data['city'], case=False, na=False)]
        
        # Фильтр по опыту работы
        experience_map = {
            "Нет опыта": ["Нет опыта"],
            "От 1 года до 3 лет": ["От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"],
            "От 3 до 6 лет": ["От 3 до 6 лет", "Более 6 лет"],
            "Более 6 лет": ["Более 6 лет"]
        }
        if search_data['experience'] in experience_map:
            filtered_df = filtered_df[filtered_df['experience.name'].isin(experience_map[search_data['experience']])]
        
               # Фильтр по типу занятости
        if search_data['employment_type'] != "Не важно":
            filtered_df = filtered_df[filtered_df['schedule.name'].str.contains(
                search_data['employment_type'], case=False, na=False)]
        
        # Фильтр по желаемой зарплате
        desired_salary = search_data['desired_salary']
        filtered_df = filtered_df[
            (filtered_df['salary.from'] <= desired_salary * 1.5) | 
            (filtered_df['salary.to'] >= desired_salary * 0.7)]
        
        # Если после базовых фильтров ничего не осталось, возвращаем пустой DataFrame
        if filtered_df.empty:
            return pd.DataFrame()
        
        # Ищем вакансии, похожие на желаемую должность
        # 1. Поиск по точному или частичному совпадению в названии
        similar_indices = []
        for idx, title in enumerate(filtered_df['name']):
            clean_title = self.clean_text(str(title))
            if target_job_clean in clean_title or clean_title in target_job_clean:
                similar_indices.append(idx)
                if len(similar_indices) >= 5:
                    break
        
        # 2. Если не нашли по названию, ищем по описанию
        if not similar_indices:
            for idx, desc in enumerate(filtered_df['processed_description_desc']):
                clean_desc = self.clean_text(str(desc))
                if target_job_clean in clean_desc:
                    similar_indices.append(idx)
                if len(similar_indices) >= 3:
                    break
        
        # 3. Если совсем не нашли, используем TF-IDF
        if not similar_indices:
            target_vector = self.tfidf_vectorizer.transform([target_job_clean])
            similarities = cosine_similarity(target_vector, self.tfidf_matrix)[0]
            similar_indices = np.argsort(-similarities)[:10].tolist()
        
        # Если нашли похожие вакансии, фильтруем по ним
        if similar_indices:
            filtered_df = filtered_df.iloc[similar_indices].copy()
        
        # Вычисляем сходство с желаемой должностью
        if not filtered_df.empty:
            job_texts = filtered_df['name'] + " " + filtered_df['processed_description_desc']
            job_vectors = self.tfidf_vectorizer.transform(job_texts)
            target_vector = self.tfidf_vectorizer.transform([target_job_clean])
            similarities = cosine_similarity(target_vector, job_vectors)[0]
            filtered_df['similarity'] = similarities
            filtered_df = filtered_df.sort_values('similarity', ascending=False)
        
        return filtered_df.head(10)
    
    def find_relevant_jobs_by_resume(self, search_data):
        """Поиск релевантных вакансий по тексту резюме с использованием TF-IDF и LSTM"""
        # Базовые фильтры
        filtered_df = self.vacancies_df.copy()
        
        # Фильтр по городу
        if search_data['city']:
            filtered_df = filtered_df[filtered_df['area.name_desc'].str.contains(
                search_data['city'], case=False, na=False)]
        
        # Фильтр по опыту работы
        experience_map = {
            "Нет опыта": ["Нет опыта"],
            "От 1 года до 3 лет": ["От 1 года до 3 лет", "От 3 до 6 лет", "ББолее 6 лет"],
            "От 3 до 6 лет": ["От 3 до 6 лет", "Более 6 лет"],
            "Более 6 лет": ["Более 6 лет"]
        }
        if search_data['experience'] in experience_map:
            filtered_df = filtered_df[filtered_df['experience.name'].isin(experience_map[search_data['experience']])]
        
        # Фильтр по типу занятости
        if search_data['employment_type'] != "Не важно":
            filtered_df = filtered_df[filtered_df['schedule.name'].str.contains(
                search_data['employment_type'], case=False, na=False)]
        
        # Фильтр по желаемой зарплате
        desired_salary = search_data['desired_salary']
        filtered_df = filtered_df[
            (filtered_df['salary.from'] <= desired_salary * 1.5) | 
            (filtered_df['salary.to'] >= desired_salary * 0.7)]
        
        # Если после базовых фильтров осталось слишком много вакансий, применяем TF-IDF и LSTM
        if len(filtered_df) > 0:
            # Получаем вектор резюме для TF-IDF
            resume_vector = self.tfidf_vectorizer.transform([search_data['resume_text']])
            
            # Получаем векторы вакансий
            job_vectors = self.tfidf_vectorizer.transform(filtered_df['combined_text'])
            
            # Вычисляем косинусное сходство для TF-IDF
            similarities_tfidf = cosine_similarity(resume_vector, job_vectors)[0]
            
            # Получаем предсказания категории LSTM для резюме
            resume_sequence = self.tokenizer.texts_to_sequences([search_data['resume_text']])
            resume_padded = pad_sequences(resume_sequence, maxlen=200)
            resume_category_prob = self.lstm_model.predict(resume_padded)[0]
            
            # Получаем предсказания категорий для вакансий
            job_sequences = self.tokenizer.texts_to_sequences(filtered_df['combined_text'].tolist())
            job_padded = pad_sequences(job_sequences, maxlen=200)
            job_categories_prob = self.lstm_model.predict(job_padded)
            
            # Вычисляем сходство категорий (косинуссное сходство между векторами вероятностей)
            similarities_lstm = cosine_similarity(
                resume_category_prob.reshape(1, -1), 
                job_categories_prob)[0]
            
            # Комбинируем оценки (можно настроить веса)
            combined_similarity = 0.6 * similarities_tfidf + 0.4 * similarities_lstm
            filtered_df['similarity'] = combined_similarity
            filtered_df = filtered_df.sort_values('similarity', ascending=False)
            
            # Ограничиваем количество результатов
            filtered_df = filtered_df.head(10)
        
        return filtered_df
    
    def display_results(self, filtered_jobs):
        """Отображение результатов поиска"""
        if filtered_jobs.empty:
            messagebox.showinfo("Результаты поиска", "К сожалению, не найдено подходящих вакансий.")
        else:
            # Создаем окно с результатами
            results_window = tk.Toplevel(self.root)
            results_window.title("Результаты поиска вакансий")
            results_window.geometry("900x600")
            
            # Создаем Treeview для отображения таблицы
            tree = ttk.Treeview(results_window, columns=('name', 'company', 'salary', 'experience', 'similarity'), show='headings')
            tree.heading('name', text='Должность')
            tree.heading('company', text='Компания')
            tree.heading('salary', text='Зарплата')
            tree.heading('experience', text='Опыт')
            tree.heading('similarity', text='Сходство')
            
            tree.column('name', width=200)
            tree.column('company', width=150)
            tree.column('salary', width=100)
            tree.column('experience', width=100)
            tree.column('similarity', width=80)
            
            # Добавляем данные
            for _, row in filtered_jobs.iterrows():
                salary_from = row['salary.from']
                salary_to = row['salary.to']
                
                if pd.notna(salary_from) and pd.notna(salary_to):
                    salary_str = f"{int(salary_from)}-{int(salary_to)}"
                elif pd.notna(salary_from):
                    salary_str = f"от {int(salary_from)}"
                elif pd.notna(salary_to):
                    salary_str = f"до {int(salary_to)}"
                else:
                    salary_str = "Не указано"
                
                similarity = row.get('similarity', 0)
                similarity_str = f"{similarity:.2f}" if isinstance(similarity, (float, np.float64)) else "N/A"
                
                tree.insert('', tk.END, values=(
                    row.get('name', 'Не указано'),
                    row.get('employer.name', 'Не указано'),
                    salary_str,
                    row.get('experience.name', 'Не указано'),
                    similarity_str))
            
            # Добавляем полосу прокрутки
            scrollbar = ttk.Scrollbar(results_window, orient="vertical", command=tree.yview)
            tree.configure(yscrollcommand=scrollbar.set)
            scrollbar.pack(side="right", fill="y")
            tree.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)
            
            # Кнопка закрытия
            ttk.Button(results_window, text="Закрыть", command=results_window.destroy).pack(pady=10)


if __name__ == "__main__":
    # Загрузка реального датасета
    try:
        # Пытаемся загрузить данные из CSV файла
        df_vacancies = pd.read_csv('vacancies.csv', low_memory=False)
        
        # Если в данных есть JSON-столбцы, преобразуем их
        json_columns = ['employment', 'experience', 'salary']
        for col in json_columns:
            if col in df_vacancies.columns:
                df_vacancies[col] = df_vacancies[col].apply(
                    lambda x: json.loads(x.replace("'", "\"")) if isinstance(x, str) else x)
        
        # Извлекаем вложенные данные
        if 'experience' in df_vacancies.columns:
            df_vacancies['experience.name'] = df_vacancies['experience'].apply(
                lambda x: x.get('name', 'Нет опыта') if isinstance(x, dict) else 'Нет опыта')
        
        if 'employment' in df_vacancies.columns:
            df_vacancies['schedule.name'] = df_vacancies['employment'].apply(
                lambda x: x.get('name', '') if isinstance(x, dict) else ''
            )
        
        if 'salary' in df_vacancies.columns:
            df_vacancies['salary.from'] = df_vacancies['salary'].apply(
                lambda x: x.get('from', None) if isinstance(x, dict) else None
            )
            df_vacancies['salary.to'] = df_vacancies['salary'].apply(
                lambda x: x.get('to', None) if isinstance(x, dict) else None
            )
            df_vacancies['salary.currency'] = df_vacancies['salary'].apply(
                lambda x: x.get('currency', 'RUR') if isinstance(x, dict) else 'RUR'
            )
        
    except Exception as e:
        print(f"Ошибка при загрузке датасета: {e}")
        # Если не удалось загрузить, используем демо-данные
        sample_data = {
            'name': ['Разработчик Python', 'Аналитик данных', 'Менеджер проектов'],
            'description': ['Разработка на Python и Django', 'Анализ данных с использованием Python и SQL', 'Управление IT проектами'],
            'salary.from': [100000, 80000, 120000],
            'salary.to': [150000, 120000, 180000],
            'experience.name': ['От 1 года до 3 лет', 'От 3 до 6 лет', 'Более 6 лет'],
            'area.name': ['Москва', 'Санкт-Петербург', 'Москва'],
            'schedule.name': ['Полный день', 'Полный день', 'Удаленная работа'],
            'alternate_url': ['http://example.com/1', 'http://example.com/2', 'http://example.com/3'],
            'employer.name': ['Компания А', 'Компания Б', 'Компания В']
        }
        
        df_vacancies = pd.DataFrame(sample_data)
    
    root = tk.Tk()
    app = JobSearchSystem(root, df_vacancies)
    root.mainloop()
```
